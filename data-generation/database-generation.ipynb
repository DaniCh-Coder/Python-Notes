```python
import pandas as pd
from mimesis import Person, Address
from mimesis.locales import Locale

# Crear instancias de Person y Address
person = Person(Locale.EN)
address = Address(Locale.EN)

# Generar un ID de cliente único y secuencial
def generar_id_cliente(index):
    return f"{index + 100000:06}"

# Lista para almacenar los datos de los clientes
clientes_data = []

# Generar 100 registros de clientes
for i in range(100):
    cliente = {
        "id_cliente": generar_id_cliente(i),
        "nombre1": person.first_name(),
        "nombre2": person.first_name(),
        "apellido": person.last_name(),
        "email": person.email(),
        "telefono": person.telephone(),
        "calle": address.street_name(),
        "numero": address.street_number(),
        "localidad": address.city(),
        "provincia": address.state(),
        "pais": address.country()
    }
    clientes_data.append(cliente)

# Convertir a un DataFrame de pandas
df_clientes = pd.DataFrame(clientes_data)
df_clientes
```

```python
# Guardar en un archivo CSV
path='../data/'
csv_file = path + 'clientes_mimesis.csv'
df_clientes.to_csv(csv_file, index=False, encoding='utf-8')

# Guardar en un archivo Excel
excel_file = path + 'clientes_mimesis.xlsx'
df_clientes.to_excel(excel_file, index=False)

print(f"Archivos '{csv_file}' y '{excel_file}' generados con éxito.")
```
