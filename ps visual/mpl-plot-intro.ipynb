```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Crear un conjunto de datos de ejemplo
data = {'category': ['A', 'B', 'C', 'D'],
        'value': [4, 7, 1, 8]}

df = pd.DataFrame(data)

# Crear el gráfico de barras para establecer la posición de los puntos
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='category', y='value', data=df, color='skyblue', width= 0.05, errorbar=None)

# Remover las barras y agregar líneas verticales y puntos en la parte superior
for index, row in df.iterrows():
    plt.plot([index, index], [0, row['value']], color='blue', linestyle='-', linewidth=2)  # Línea vertical
    plt.plot(index, row['value'], 'bo')  # Punto en la parte superior

# Ajustar las etiquetas del eje x y mover el eje x a la posición 0
plt.xticks(range(len(df)), df['category'])
plt.axhline(0, color='black', linewidth=1)  # Línea del eje x en y=0

# Ajustar las etiquetas y título
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Gráfico de líneas con puntos en la parte superior')

# Mostrar el gráfico
plt.show()


```
