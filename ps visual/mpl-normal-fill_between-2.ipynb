```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import math
```

```python
plt.figure(figsize=(6,5))
plt.fill_between(x=np.arange(0,10,0.01), 
                 y1= stats.norm.pdf(np.arange(0,10,0.01), loc = 5), 
                 facecolor='grey',
                 alpha=0.35)
plt.fill_between(x=np.arange(6,10,0.01), 
                 y1= stats.norm.pdf(np.arange(6,10,0.01), loc = 5) ,
                 facecolor='red',
                 alpha=0.5)
```
