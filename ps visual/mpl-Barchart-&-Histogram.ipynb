# Grafico de barras simple
##### Ejemplo de distribución de edades de actrices ganadoras del oscar en gráfico de barras.
###### Observese que el grafico toma los datos de una tabla de distribución de frecuencias ya dada (rango_edades, frecuencia).
###### Es decir, no toma los 76 datos en bruto de las edades. 

```python
import matplotlib.pyplot as plt

# Datos
rango_edades = ['21-30', '31-40', '41-50', '51-60', '61-70', '71-80']
frecuencia = [28, 30, 12, 2, 2, 2]

# Crear el gráfico de barras
plt.bar(rango_edades, frecuencia, color='skyblue')

# Añadir títulos y etiquetas
plt.title('Distribución de Frecuencia por Rango de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```

# Histograma
##### Ejemplo de distribución de edades de actrices ganadoras del oscar en un histograma.
###### Observese que el gráfico toma los 76 datos del conjunto de edades observadas.
###### Es decir no toma los datos de una tabla de distribución de frecuencias ya armada. (rango_edades, frecuencia)

```python
import matplotlib.pyplot as plt

# Datos
edades = []
edades.extend([25] * 28)  # Aproximando edades en el rango 21-30
edades.extend([35] * 30)  # Aproximando edades en el rango 31-40
edades.extend([45] * 12)  # Aproximando edades en el rango 41-50
edades.extend([55] * 2)   # Aproximando edades en el rango 51-60
edades.extend([65] * 2)   # Aproximando edades en el rango 61-70
edades.extend([75] * 2)   # Aproximando edades en el rango 71-80
print(edades)
# Crear el histograma
plt.hist(edades, bins=[20, 30, 40, 50, 60, 70, 80], color='skyblue', edgecolor='gray', rwidth=1.0)

# Añadir títulos y etiquetas
plt.title('Distribución de Frecuencia por Rango de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')

# Mostrar el gráfico
plt.show()
```

```python
# Datos
edades = []
edades.extend([25] * 28)  # Aproximando edades en el rango 21-30
edades.extend([35] * 30)  # Aproximando edades en el rango 31-40
edades.extend([45] * 12)  # Aproximando edades en el rango 41-50
edades.extend([55] * 2)   # Aproximando edades en el rango 51-60
edades.extend([65] * 2)   # Aproximando edades en el rango 61-70
edades.extend([75] * 2)   # Aproximando edades en el rango 71-80

# Crear el histograma
counts, bins, patches = plt.hist(edades, bins=[20, 30, 40, 50, 60, 70, 80], color='skyblue', edgecolor='gray', rwidth=1.0, density=True)

# Añadir títulos y etiquetas
plt.title('Frecuencia Relativa por Edad (en %)')
plt.xlabel('Rango de Edad')
# Eliminar el eje vertical (eje Y)
plt.gca().axes.yaxis.set_visible(False)
# Eliminar el borde superior, izquierdo y derecho
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Añadir valores de frecuencia relativa en la parte superior de cada barra
for count, patch in zip(counts, patches):       # itera sobre cada barra del histograma.
    height = patch.get_height()                 # patch.get_height() obtiene la altura de la barra actual.
    plt.annotate(f'{1000 * height:2.2f}',                # mostrar height (altura) cuatro decimales
                 xy=(patch.get_x() + patch.get_width() / 2, height),  # xy define la posición del texto (el centro superior de la barra).
                 xytext=(0, 5),                 # Desplazamiento del texto (en puntos)
                 textcoords='offset points',    # indica que el desplazamiento se da en puntos.
                 ha='center', va='bottom')      # alinean el texto horizontal y verticalmente.

# Mostrar el gráfico
plt.show()
```
