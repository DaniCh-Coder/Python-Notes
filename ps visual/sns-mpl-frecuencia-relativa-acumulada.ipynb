```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos proporcionados
data = {
    'Edad': ['3539', '4044', '4549', '5054', '5559', '6064', '6569'],
    'Frecuencia': [1, 3, 5, 11, 7, 7, 1]
}
df = pd.DataFrame(data)

# Crear la distribucin de frecuencia simple
plt.figure(figsize=(10, 5))
sns.barplot(x='Edad', y='Frecuencia', data=df, color='skyblue')
plt.title('Distribucin de Frecuencia Simple')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')
plt.show()

# Calcular la distribucin de frecuencia acumulada
df['Frecuencia Acumulada'] = df['Frecuencia'].cumsum()

# Crear la distribucin de frecuencia acumulada
plt.figure(figsize=(10, 5))
sns.barplot(x='Edad', y='Frecuencia Acumulada', data=df, color='salmon')
plt.title('Distribucin de Frecuencia Acumulada')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia Acumulada')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos proporcionados
data = {
    'Edad': ['3539', '4044', '4549', '5054', '5559', '6064', '6569'],
    'Frecuencia': [1, 3, 5, 11, 7, 7, 1]
}
df = pd.DataFrame(data)

# Calcular la distribucin de frecuencia acumulada
df['Frecuencia Acumulada'] = df['Frecuencia'].cumsum()

# Crear el grfico combinado de frecuencia simple y acumulada
plt.figure(figsize=(10, 5))
sns.barplot(x='Edad', y='Frecuencia', data=df, color='skyblue', label='Frecuencia Simple')
sns.lineplot(x='Edad', y='Frecuencia Acumulada', data=df, color='salmon', marker='o', label='Frecuencia Acumulada')
plt.title('Distribucin de Frecuencia Simple y Acumulada')
plt.xlabel('Rango de Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos proporcionados
data = {
    'Edad': ['3539', '4044', '4549', '5054', '5559', '6064', '6569'],
    'Frecuencia': [1, 3, 5, 11, 7, 7, 1]
}
df = pd.DataFrame(data)

# Calcular la distribucin de frecuencia acumulada
df['Frecuencia Acumulada'] = df['Frecuencia'].cumsum()

# Crear un solo plot con dos subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle('Distribucin de Frecuencias', fontsize=14, ha='center')

# Subplot para la distribucin de frecuencia simple
axs[0].bar(df['Edad'], df['Frecuencia'], color='skyblue')
axs[0].set_title('Simple')
axs[0].set_xlabel('Rango de Edad')
axs[0].set_ylabel('Frecuencia')
axs[0].tick_params(left=False, bottom=False)  # Quitar los ejes y rtulos

# Agregar valores en la parte central de cada barra (distribucin de frecuencia simple)
for index, row in df.iterrows():
    axs[0].text(index, row['Frecuencia'] / 2, str(row['Frecuencia']), ha='center', va='center', color='black')

# Subplot para la distribucin de frecuencia acumulada
axs[1].bar(df['Edad'], df['Frecuencia Acumulada'], color='salmon')
axs[1].set_title('Acumulada')
axs[1].set_xlabel('Rango de Edad')
axs[1].set_ylabel('Frecuencia Acumulada')
axs[1].tick_params(left=False, bottom=False)  # Quitar los ejes y rtulos

# Agregar valores en la parte central de cada barra (distribucin de frecuencia acumulada)
for index, row in df.iterrows():
    axs[1].text(index, row['Frecuencia Acumulada'] / 2, str(row['Frecuencia Acumulada']), ha='center', va='center', color='black')

# Ajustar el diseo de los subplots
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Ajustar el espacio entre los subplots y el ttulo principal

# Mostrar el plot
plt.show()
```
