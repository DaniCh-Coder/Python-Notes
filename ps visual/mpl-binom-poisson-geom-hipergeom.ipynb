## Generaci贸n de distribuciones por simulaci贸n
### Binomial, Poisson, Gemo茅trica, Hipergeom茅trica

```python
import numpy as np  # generaci贸n aleatoria de numeros
import matplotlib.pyplot as plt # generaci贸n de gr谩ficos
```

### Funci贸n para graficar las distintas distribuciones de probabilidad

```python
def DibujarDistribucion(nombre= "", samples = None, bins = 1):
    titulo = "Distribuci贸n " + nombre    
    # Crear un solo plot con dos subplots
    plt.figure(figsize=(12, 6))
    plt.suptitle(titulo, fontsize=14, ha='center')

    # Subplot para la distribuci贸n de frecuencia simple
    plt.subplot(1, 2, 1)
    plt.hist(samples, bins)
    plt.title('Simple')

    # Subplot para la distribuci贸n de probabilidad
    plt.subplot(1, 2, 2)
    plt.hist(samples, bins, density=True)
    plt.title('Probabilidad')
    plt.xlabel('Rango de Edad')
```

# Distribuciones de probabilidad
## Distribuci贸n Binomial
La distribuci贸n binomial describe el n煤mero de 茅xitos en una secuencia de  experimentos de Bernoulli independientes, 
+ cada uno con una probabilidad de 茅xito .

```python

# Par谩metros
n = 10    # N煤mero de ensayos
p = 0.5   # Probabilidad de 茅xito en cada ensayo
size = 1000  # N煤mero de simulaciones
# cada muestra (size = 1000) es un experimento de (n=10) ensayos
# cada ensayo tiene una probabilidad d 0.5 de 茅xito

# Simulaci贸n
binomial_samples = np.random.binomial(n, p, size)
bins = n+1

# Histograma
DibujarDistribucion("Binomial", binomial_samples, bins)
```

## Distribuci贸n de Poisson
La distribuci贸n de Poisson describe el n煤mero de eventos que ocurren en un intervalo de tiempo fijo o espacio, con una tasa promedio .

```python
# Par谩metro
lam = 3  # Tasa promedio de eventos
size = 1000  # N煤mero de simulaciones

# Simulaci贸n
poisson_samples = np.random.poisson(lam, size)
bins = range(0, max(poisson_samples)+1)

# Histograma
DibujarDistribucion("Poisson", poisson_samples, bins)
```

## Distribuci贸n Geom茅trica
La distribuci贸n geom茅trica describe el n煤mero de ensayos necesarios para obtener el primer 茅xito en una serie de experimentos de Bernoulli.

```python
# Par谩metro
p = 0.5  # Probabilidad de 茅xito en cada ensayo
size = 1000  # N煤mero de simulaciones

# Simulaci贸n
geometric_samples = np.random.geometric(p, size)
bins=range(1, max(geometric_samples)+1)

# Histograma
DibujarDistribucion("Geom茅trica", geometric_samples, bins)
```

## Distribuci贸n Hipergeom茅trica
La distribuci贸n hipergeom茅trica describe el n煤mero de 茅xitos en una muestra de tama帽o  extra铆da sin reemplazo
+ de una poblaci贸n de tama帽o  que contiene  茅xitos.

```python
# Par谩metros
N = 20   # Tama帽o de la poblaci贸n
K = 7    # N煤mero de 茅xitos en la poblaci贸n
n = 5    # Tama帽o de la muestra
size = 1000  # N煤mero de simulaciones

# Simulaci贸n
hypergeometric_samples = np.random.hypergeometric(K, N - K, n, size)
bins=range(0, n+1)

# Histograma
DibujarDistribucion("Hipergeom茅trica", hypergeometric_samples, bins)
```
