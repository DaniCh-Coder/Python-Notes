# Distribuciones de probabilidad Discreta
## Generación de distribuciones por simulación
### Binomial, Poisson, Gemoétrica, Hipergeométrica

```python
import numpy as np  # generación aleatoria de numeros
import matplotlib.pyplot as plt # generación de gráficos
```

### Función para graficar las distintas distribuciones de probabilidad

```python
def DibujarDistribucion(nombre= "", samples = None, bins = 1):
    titulo = "Distribución " + nombre    
    # Crear un solo plot con dos subplots
    plt.figure(figsize=(12, 6))
    plt.suptitle(titulo, fontsize=14, ha='center')

    # Subplot para la distribución de frecuencia simple
    plt.subplot(1, 2, 1)
    plt.hist(samples, bins)
    plt.title('Simple')

    # Subplot para la distribución de probabilidad
    plt.subplot(1, 2, 2)
    plt.hist(samples, bins, density=True)
    plt.title('Probabilidad')
    plt.xlabel('Rango de Edad')
```

# Distribuciones de probabilidad
## Distribución Binomial
La distribución binomial describe el número de éxitos en una secuencia de 𝑛 experimentos de Bernoulli independientes, 
+ cada uno con una probabilidad de éxito 𝑝.

```python

# Parámetros
n = 10    # Número de ensayos
p = 0.5   # Probabilidad de éxito en cada ensayo
size = 1000  # Número de simulaciones
# cada muestra (size = 1000) es un experimento de (n=10) ensayos
# cada ensayo tiene una probabilidad d 0.5 de éxito

# Simulación
binomial_samples = np.random.binomial(n, p, size)
bins = n+1

# Histograma
DibujarDistribucion("Binomial", binomial_samples, bins)
```

## Distribución de Poisson
La distribución de Poisson describe el número de eventos que ocurren en un intervalo de tiempo fijo o espacio, con una tasa promedio 𝜆.

```python
# Parámetro
lam = 3  # Tasa promedio de eventos
size = 1000  # Número de simulaciones

# Simulación
poisson_samples = np.random.poisson(lam, size)
bins = range(0, max(poisson_samples)+1)

# Histograma
DibujarDistribucion("Poisson", poisson_samples, bins)
```

## Distribución Geométrica
La distribución geométrica describe el número de ensayos necesarios para obtener el primer éxito en una serie de experimentos de Bernoulli.

```python
# Parámetro
p = 0.5  # Probabilidad de éxito en cada ensayo
size = 1000  # Número de simulaciones

# Simulación
geometric_samples = np.random.geometric(p, size)
bins=range(1, max(geometric_samples)+1)

# Histograma
DibujarDistribucion("Geométrica", geometric_samples, bins)
```

## Distribución Hipergeométrica
La distribución hipergeométrica describe el número de éxitos en una muestra de tamaño 𝑛 extraída sin reemplazo
+ de una población de tamaño 𝑁 que contiene 𝐾 éxitos.

```python
# Parámetros
N = 20   # Tamaño de la población
K = 7    # Número de éxitos en la población
n = 5    # Tamaño de la muestra
size = 1000  # Número de simulaciones

# Simulación
hypergeometric_samples = np.random.hypergeometric(K, N - K, n, size)
bins=range(0, n+1)

# Histograma
DibujarDistribucion("Hipergeométrica", hypergeometric_samples, bins)
```
