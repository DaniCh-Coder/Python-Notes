2) Split the only column into multiple columns on a temporal dataframe

```python
# create a temporal data frame
pdTemp = pd.DataFrame()

# take each row in the file dataframe (df) and split it in columns
# separate columns by space
for i in range(170):
    row = pd.Series(df.iloc[i,0].split())
    pdTemp = pd.concat([pdTemp, row.to_frame().T], ignore_index=True)

# see the result
pdTemp
```

3) Process the column names
    - Edit the column names stored in var columnames matchimg the 6 columns in pdTemp
    - Set the column axe to editated names in columnames
    - Once it is Ok, first row (in df.iloc[0, :]) can be deleted 

Now restore the temporal dataframe pdTemp into the definitive dataframe df

```python
# Edit columnames variable, based on observations of pdTemp
columnames=['x','Duration', 'Date', 'Pulse', 'Maxpulse', 'Calories']

# Set new names of columns, seting axis=1
df = pdTemp.set_axis(columnames,axis=1)

# Eliminate row 0 because it contains the name of columns
df.drop([0], inplace=True)
df
```

4) Observations: See that we are improved the file. But we still have issues.
    - We do not need the column 'x'. It can be dropped.
    - Check if column names match with columns and it make sense
    - Eliminate the symbol";" from the las column in the rows that have it

```python
# Drop unnecesary 'x' column
df.drop(['x'], axis=1, inplace=True)

# Eliminate ";" form the last column in some rows
df['Calories'] = df['Calories'].str.replace(';','')
df
```

5) Observations: Better. More improvements needed.
    - First part of the dataframe is far more better.
    - Last part of the dataframe has some mess in 'Maxpulse' column:
        - It's seen that it is something like:
            - calories, row, duration, pulse, max pulse, calories
            - proccess 'Maxpulse' it's necesary

6) Proccess 'Maxpulse' column:
    - Split the values on columns in a temporal dataframe
    - Retrieve necesary values
    - Discard unnecesary columns


```python
# create a temporal dataframe
dft = pd.DataFrame()

# create a list to sotre the rows with this issue
positions = []

# split the each row of 'Maxpluse'. Sotre the columns in dft
for i, value in enumerate(df['Maxpulse']):
    if ';' in value:
        row = pd.Series(value.split(';'))
        dft = pd.concat([dft, row.to_frame().T], ignore_index=True)
        positions += [i+1]

# take the first row with the problem
firstpos = positions[0]

# here it is a sub-array of the main dataframe
dft.head(5)
```

7) Observation: unnecesary columns analysis and reindex the results

```python
# make a column analisis. set column names hipothesis. check results
# (we are depurating a subarray of the main dataframe)
colanalysis=['x1', 'x2', 'Duration', 'Pulse', 'Maxpulse', 'Calories','x3']
dft = dft.set_axis(colanalysis,axis=1)
dft.head(2)
```

```python

# drop unnecesary columns
dft.drop(['x1','x2','x3'], axis=1, inplace=True)
```

```python

# reset index with original dataframe df number of rows
# for further update
dft.set_index(pd.Series(positions), inplace=True)
dft
```

8) Update the main dataframe (df) with the depurated portion (dft)

```python
# locate the main dataframe portion to be updated with corrections
# Once located, update
df.loc[firstpos:,['Duration', 'Pulse', 'Maxpulse','Calories']] = dft

# check the results
df
```

9) Erease dates that are not dates
