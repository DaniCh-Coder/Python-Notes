## Clean and convert to numbers

### Eliminate invisible characters

#### Try to replace " "
Note: It's seen that not all invisible characters are eliminated witha simple replace.

```python
df.Asia.str.replace(" ", "")
```

#### Try to replace all invisible characters
the way to eliminate all invisible characters is with: series.str.replace(r'\s+', '', regex=True)

```python
df.Asia.str.replace(r'\s+', '', regex=True)
```

Here's a breakdown of the code series.str.replace(r'\s+', '', regex=True):

1. series.str.replace(...): 
    + series.str: Accesses string-specific methods for the pandas Series.
    + .replace(...): Replaces occurrences of a pattern with a specified replacement.

2. r'\s+':
    + r: Prefixes the string to make it a raw string, ensuring backslashes are interpreted literally (not as escape characters).
    + \s: Regular expression pattern matching any whitespace character, including:
        * Regular spaces
        * Non-breaking spaces
        * Tabs
        * Newlines
    + +: Quantifier meaning "one or more times," so \s+ matches one or more consecutive whitespace characters.

3. '': Empty string used as the replacement, effectively removing the matched whitespace characters.

4. regex=True: Flag indicating that the first argument (r'\s+') should be treated as a regular expression pattern, enabling pattern matching functionality.

##  Replace invisible chars and convert to number

```python
for col in df.columns:
    df[col] = df[col].str.replace(r'\s+', '', regex=True).astype('int64')
```
