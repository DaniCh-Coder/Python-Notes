```python
from collections import Counter
from wordcloud import WordCloud
```

1. Open the document to analyze. 2. Consider to prepare and open the stopwords file specially if the text is in other languaje.


```python
# Open the file to make the wordcloud
with open("DNU70-2023-Argentina.txt", encoding="utf-8") as f:
    dnutxt = f.read()

# Open the file with most common stopwords in spanish
with open("stoppalabras.txt", encoding="utf-8") as f:
    stoppalabras = f.read()
    stoppal = stoppalabras.split('\n')
```

3. Analyze the frecuency of the words in the text to work with

```python

# Calculate occurences of words
# create a dictionary with dict() to count repetitions of each number
# using Counter() from collections
d = dict(Counter(dnutxt.split()))
```

```python
# Order word by frecuency of occurence in descending order
d_orden = dict(sorted(d.items(), key=lambda item: item[1], reverse=True))
```

```python
# Take the most frequent words (in d_orden), filter them with the most common words (in stoppalabras)
n_lista = { k : v for k, v in d_orden.items() if k not in stoppal and v > 120 }
```

```python
# Analyze the result. Consider if it is necesary to add some other not meaningfull word to stop
n_lista
```

3. Construct the definitive stopwords list

```python
# In this case see no signigicant words like: 'N°', 'artículo', ARTÍCULO', “ARTÍCULO', 'siguiente'
# Extend this words to the sotppal list
stoppal.extend(['N°', 'artículo', '“ARTÍCULO', 'siguiente', 'N', 'N °', 'e'])
# set filter if there is duplicates. Also WordCloudClass take a set() as stopwords arg
stopw = set(stoppal)
```

4. Start working in canvas wordclaud

```python
# Generate a word cloud image with text and stopwords
wordcloud = WordCloud(stopwords=stopw).generate(dnutxt)
```

```python
# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
# store to file
wordcloud.to_file("DNU_400X200.png")
```
