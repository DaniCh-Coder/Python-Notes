# Beautiful Soup - Open a File

```python
import requests
import html5lib
import xml
from bs4 import BeautifulSoup
```

```python
# Fetch the HTML content from the web page
response = requests.get('https://www.ambito.com/contenidos/dolar-informal-historico.html')
# Check for successful response
response.raise_for_status()  # Raises an exception for error status codes
```

```python
# Create a BeautifulSoup object from the HTML content
soup = BeautifulSoup(response.content)
```

```python
soup.prettify()
```

```python
# Find tables
tables = soup.find_all("table")
```

```python
# find tables in the file
for table in tables:
    print(table)
```

```python
for child in table:
    print(child)
```

```python
tag_head = table.thead
tag_head
```

```python
tag_tr = tag_head.children
tag_tr
```

```python
for i in tag_tr:
    print(i)
```
