#### Ejemplo de uso de la librer√≠a pymc

```python
import pymc as pm
import numpy as np

# Datos de ejemplo
datos = np.random.binomial(1, 0.5, size=100)

with pm.Model() as modelo_bayesiano:
    # Paso 1: Definir la distribuci√≥n a priori para p
    p = pm.Beta('p', alpha=1, beta=1)
    
    # Paso 2: Definir la verosimilitud en funci√≥n de los datos observados
    observaciones = pm.Binomial('observaciones', n=1, p=p, observed=datos)
    
    # Paso 3: Muestreo de la distribuci√≥n posterior
    posterior = pm.sample(1000)
    
    # Paso 4: Graficar la distribuci√≥n posterior
    pm.plot_posterior(posterior)
```
