```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parmetros
alpha = 0.05  # Nivel de significancia
n = 30  # Tamao de la muestra
df = n - 1  # Grados de libertad
mu = 100  # Media poblacional
sigma = 15  # Desviacin estndar poblacional
sample_mean = 108  # Media de la muestra

# Valor crtico para una prueba de dos colas
t_critical = stats.t.ppf(1 - alpha/2, df)

# Regin crtica
critical_region_left = mu - t_critical * (sigma / np.sqrt(n))
critical_region_right = mu + t_critical * (sigma / np.sqrt(n))

# Estadstico t de la muestra
t_statistic = (sample_mean - mu) / (sigma / np.sqrt(n))

# Grfico
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma / np.sqrt(n))

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribucin de la muestra')
plt.axvline(critical_region_left, color='red', linestyle='dashed', label='Valor crtico izquierdo')
plt.axvline(critical_region_right, color='red', linestyle='dashed', label='Valor crtico derecho')
plt.axvline(sample_mean, color='blue', linestyle='dashed', label='Media de la muestra')
plt.fill_between(x, y, where=(x < critical_region_left) | (x > critical_region_right), color='red', alpha=0.2, label='Regin crtica')
plt.legend()
plt.title('Valor Crtico y Regin Crtica en una Prueba de Hiptesis')
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f'Valor crtico izquierdo: {critical_region_left:.2f}')
print(f'Valor crtico derecho: {critical_region_right:.2f}')
print(f'Estadstico t de la muestra: {t_statistic:.2f}')

# Interpretacin
if t_statistic < -t_critical or t_statistic > t_critical:
    print("Rechazamos la hiptesis nula. El valor cae en la regin crtica.")
else:
    print("No podemos rechazar la hiptesis nula. El valor no cae en la regin crtica.")
```
