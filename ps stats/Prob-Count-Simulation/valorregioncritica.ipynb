```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros
alpha = 0.05  # Nivel de significancia
n = 30  # Tamaño de la muestra
df = n - 1  # Grados de libertad
mu = 100  # Media poblacional
sigma = 15  # Desviación estándar poblacional
sample_mean = 108  # Media de la muestra

# Valor crítico para una prueba de dos colas
t_critical = stats.t.ppf(1 - alpha/2, df)

# Región crítica
critical_region_left = mu - t_critical * (sigma / np.sqrt(n))
critical_region_right = mu + t_critical * (sigma / np.sqrt(n))

# Estadístico t de la muestra
t_statistic = (sample_mean - mu) / (sigma / np.sqrt(n))

# Gráfico
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma / np.sqrt(n))

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución de la muestra')
plt.axvline(critical_region_left, color='red', linestyle='dashed', label='Valor crítico izquierdo')
plt.axvline(critical_region_right, color='red', linestyle='dashed', label='Valor crítico derecho')
plt.axvline(sample_mean, color='blue', linestyle='dashed', label='Media de la muestra')
plt.fill_between(x, y, where=(x < critical_region_left) | (x > critical_region_right), color='red', alpha=0.2, label='Región crítica')
plt.legend()
plt.title('Valor Crítico y Región Crítica en una Prueba de Hipótesis')
plt.xlabel('Valores')
plt.ylabel('Densidad de probabilidad')
plt.show()

# Resultados
print(f'Valor crítico izquierdo: {critical_region_left:.2f}')
print(f'Valor crítico derecho: {critical_region_right:.2f}')
print(f'Estadístico t de la muestra: {t_statistic:.2f}')

# Interpretación
if t_statistic < -t_critical or t_statistic > t_critical:
    print("Rechazamos la hipótesis nula. El valor cae en la región crítica.")
else:
    print("No podemos rechazar la hipótesis nula. El valor no cae en la región crítica.")
```
