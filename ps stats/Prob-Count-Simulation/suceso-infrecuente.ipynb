# Suceso infrecuente

1. Definición: Un suceso infrecuente se considera aquel que tiene una probabilidad muy baja de ocurrir, generalmente menos del 5%.

2. Aplicación: Esta regla es usada para realizar inferencias estadísticas, como en pruebas de hipótesis. 
    + Por ejemplo, si al realizar un experimento, los resultados observados tienen una probabilidad menor al 5%, 
        + podemos considerar que esos resultados no son debidos al azar.

3. Ejemplo práctico: Supongamos que estamos lanzando una moneda y deseamos determinar si la moneda está sesgada. 
    + Si lanzamos la moneda 100 veces y obtenemos 95 caras, podemos usar la regla del suceso infrecuente para evaluar 
        + si este resultado es tan raro como para considerar que la moneda no es justa.

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Parámetros
n_trials = 100
p = 0.5  # Probabilidad de obtener cara en una moneda justa
observed_heads = 95  # Número de caras observadas

# Distribución binomial
x = np.arange(0, n_trials+1)
binom_dist = stats.binom.pmf(x, n_trials, p)

# Probabilidad de obtener 95 o más caras
prob = 1 - stats.binom.cdf(observed_heads - 1, n_trials, p)

# Gráfico
plt.bar(x, binom_dist, label='Distribución binomial')
plt.axvline(observed_heads, color='red', linestyle='dashed', label=f'Observación (95 caras)')
plt.legend()
plt.title('Distribución de probabilidad binomial para 100 lanzamientos de moneda')
plt.xlabel('Número de caras')
plt.ylabel('Probabilidad')
plt.show()

print(f'La probabilidad de obtener {observed_heads} o más caras en {n_trials} lanzamientos es {prob:.5f}')
```
