```python
import itertools

# Example data
data = [0,1]

## Permutations without repetitions
print("Permutations without repetitions:")
# Create permutations of all elements
permutations = itertools.permutations(data, 2)

# Print each permutation without repetition
for permutation in permutations:
    print(permutation)

## Permutations with repetitions or replacement
print("Permutation with repetitions:")
permutationwrepeat = itertools.product(data, repeat=2)

# Print each permutation
for permutation in permutationwrepeat:
    print(permutation)

## Combinations
print("Combinations without repetitions:")   
# Create conmutations of all elements
combinations = itertools.combinations(data, 2)

# Print each combination
for combination in combinations:
    print(combination)
```

```python
import pandas as pd
# Create permutations of elements in data (ej 1, 2, 5) grouped by 2
# It also counts how many odd element are in each permutation.
# After, it print each permutation adding the number of odds in each tuple
# Permutations are with repetitions or replacement

# data
data=[1,2,5]
print("Permutation with repetitions of {data} grouped by 2:")

# Create permutation with replacement
permutationwrepeat = itertools.product(data, repeat=2)

rows= []
# Print each permutation whith replacement and the quantity of odds
for i, permutation in enumerate(permutationwrepeat):
    impares = sum(num % 2 != 0 for num in permutation)
    rows.append({'index': i, 'permutation': permutation, 'odd_count': impares})    

# Create DataFrame
df = pd.DataFrame(rows, columns=['index', 'permutation', 'odd_count'])
df.set_index('index', inplace=True)
# df.reset_index(inplace=True)
df
```
