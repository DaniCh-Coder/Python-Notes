# Beta y potencia de una muestra

```python
import numpy as np
import statsmodels.stats.power as smp
import matplotlib.pyplot as plt

# Parámetros
efecto_tamaño = 0.5  # Tamaño del efecto
n_muestra = np.arange(5, 101)  # Tamaños de la muestra de 5 a 100
alpha = 0.05  # Nivel de significancia

# Cálculo de la potencia
potencias = [smp.tt_ind_solve_power(effect_size=efecto_tamaño, nobs1=n, alpha=alpha, power=None, ratio=1) for n in n_muestra]

# Gráfico de la potencia
plt.figure(figsize=(10, 6))
plt.plot(n_muestra, potencias, marker='o')
plt.title('Potencia de la Prueba t vs. Tamaño de la Muestra')
plt.xlabel('Tamaño de la Muestra')
plt.ylabel('Potencia')
plt.axhline(y=0.8, color='r', linestyle='--', label='Potencia = 0.8')
plt.legend()
plt.grid()
plt.show()
```
