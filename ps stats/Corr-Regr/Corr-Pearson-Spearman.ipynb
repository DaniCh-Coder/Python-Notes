```python
import numpy as np
from scipy.stats import pearsonr, spearmanr
```

+ Hip√≥tesis nula (ùêª0): No existe correlaci√≥n entre las variables ùëã e ùëå. 
    + Matem√°ticamente, esto se expresa como ùëü=0 para Pearson y ùúå=0 para Spearman.

+ Hip√≥tesis alternativa (ùêª1): Existe una correlaci√≥n significativa entre las variables ùëã e ùëå. 
    + Esto se expresa como ùëü‚â†0 para Searson o ùúå‚â†0 para Spearman.

### Coeficiente de Pearson
El coeficiente de correlaci√≥n lineal de Pearson, denotado como ùëü, mide la fuerza y la direcci√≥n de la relaci√≥n lineal entre dos variables cuantitativas.

```python
# Datos de ejemplo
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)

print(f"Coeficiente de correlaci√≥n de Pearson: {r} con p_value: {p_value}")
```

**Observaci√≥n**
+ Hip√≥tesis nula (ùêª0): No existe correlaci√≥n entre las variables ùëã e ùëå. 
    + Matem√°ticamente, esto se expresa como ùëü=0 para Pearson.
+ Hip√≥tesis alternativa (ùêª1): Existe una correlaci√≥n significativa entre las variables ùëã e ùëå. 
    + Esto se expresa como ùëü‚â†0 para Pearson.
+ Valor p: Si el valor p es peque√±o (t√≠picamente ‚â§ 0.05), esto indica que la probabilidad de observar una correlaci√≥n tan fuerte como la obtenida, 
    + bajo la suposici√≥n de que ùëü=0 (no hay correlaci√≥n), es muy baja. Es decir que es muy probable que si exista una correlaci√≥n.

**Interpretaci√≥n**
+ Un valor p bajo nos lleva a rechazar la hip√≥tesis nula, sugiriendo que s√≠ hay una correlaci√≥n lineal significativa entre las variables.


### Coeficiente de Spearman
El coeficiente de correlaci√≥n de Spearman, denotado como ùúå (rho), mide la relaci√≥n mon√≥tona (no necesariamente lineal) entre dos variables cuantitativas. 
+ Se basa en los rangos de los datos, no en los valores originales, lo que lo hace adecuado para datos no param√©tricos.

```python
# Datos de ejemplo
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 60]

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)

print(f"Coeficiente de correlaci√≥n de Spearman: {rho}, valor p: {rho}")
```

**Observaci√≥n**
+ Para el Coeficiente de Correlaci√≥n de Spearman:
    + Valor p: Si el valor p es bajo, indica que es improbable observar una correlaci√≥n mon√≥tona tan fuerte como la obtenida si ùúå=0 (no hay correlaci√≥n mon√≥tona).

**Interpretaci√≥n** 
+ Un valor p bajo sugiere que hay evidencia suficiente para rechazar la hip√≥tesis nula y aceptar que existe una correlaci√≥n mon√≥tona significativa entre las variables.

### Otros 2 Ejemplos con visualizaciones

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from scipy.stats import linregress

# Generaci√≥n de datos de ejemplo
np.random.seed(42)  # Fijamos una semilla para reproducibilidad

# Datos para Pearson (relaci√≥n lineal)
x_pearson = np.linspace(10, 100, 20)
y_pearson = 3 * x_pearson + np.random.normal(0, 10, 20)  # Relaci√≥n lineal con algo de ruido

# Datos para Spearman (relaci√≥n mon√≥tona no lineal)
x_spearman = np.linspace(10, 100, 20)
y_spearman = np.log(x_spearman) * 20 + np.random.normal(0, 5, 20)  # Relaci√≥n mon√≥tona no lineal con ruido

# C√°lculo del coeficiente de Pearson
r_pearson, p_value_pearson = pearsonr(x_pearson, y_pearson)

# C√°lculo del coeficiente de Spearman
rho_spearman, p_value_spearman = spearmanr(x_spearman, y_spearman)

# Ajuste lineal para el gr√°fico de Pearson
slope_pearson, intercept_pearson, _, _, _ = linregress(x_pearson, y_pearson)
line_pearson = slope_pearson * x_pearson + intercept_pearson

# Ajuste lineal para el gr√°fico de Spearman
slope_spearman, intercept_spearman, _, _, _ = linregress(x_spearman, y_spearman)
line_spearman = slope_spearman * x_spearman + intercept_spearman

# Visualizaci√≥n de los datos, correlaciones y l√≠neas de ajuste
plt.figure(figsize=(12, 6))

# Gr√°fico para Pearson
plt.subplot(1, 2, 1)
plt.scatter(x_pearson, y_pearson, color='blue', label='Datos')
plt.plot(x_pearson, line_pearson, color='red', label=f'L√≠nea de ajuste\nr = {r_pearson:.2f}, p = {p_value_pearson:.2g}')
plt.title('Correlaci√≥n de Pearson')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

# Gr√°fico para Spearman
plt.subplot(1, 2, 2)
plt.scatter(x_spearman, y_spearman, color='green', label='Datos')
plt.plot(x_spearman, line_spearman, color='red', label=f'L√≠nea de ajuste\nrho = {rho_spearman:.2f}, p = {p_value_spearman:.2g}')
plt.title('Correlaci√≥n de Spearman')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Casos particulares. Patrones

#### Alta correlaci√≥n

```python
# Datos de ejemplo
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

# Grafico
plt.scatter(x, y, color='blue', label='Datos')

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)

print(f"Coeficiente de correlaci√≥n de Pearson: {r} con p_value: {p_value}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)

print(f"Coeficiente de correlaci√≥n de Spearman: {rho}, valor p: {rho}")
```

#### Baja correlaci√≥n

1. Examinemos un patron de datos como el siguiente

```python
# Datos de ejemplo
x = [1, 1, 1, 2, 2, 2, 3, 3, 3]
y = [1, 2, 3, 1, 2, 3, 1, 2, 3]
```

2. Observemos el diagrama de dispersi√≥n. Determinemos de manera subjetiva si existe alg√∫n grado de correlaci√≥n x,y.

```python
plt.scatter(x, y, color='blue', label='Datos')
```

+ Parece no haber correlaci√≥n porque la diagonal principal se ve absolutamente compensada por una diagonal secundaria.
+ As√≠mismo hay rectas verticales y orizontales que se compensan por lo que no se ve correlaci√≥n alguna.

3. Calculemos la correlaci√≥n para verificar nuestra observaci√≥n.

```python
# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)

print(f"Coeficiente de correlaci√≥n de Pearson: {r} con p_value: {p_value}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)

print(f"Coeficiente de correlaci√≥n de Spearman: {rho}, valor p: {rho}")
```

+ Efectivamente se verifica que no hay correlaci√≥n alguna.

### Efectos de aglomerados

```python
# Datos de ejemplo
x = [1, 1, 2, 2, 9, 9, 10, 10]
y = [1, 2, 1, 2, 9, 10, 9, 10]

plt.scatter(x, y, color='blue', label='Datos')
plt.show()

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)

print(f"Coeficiente de correlaci√≥n de Pearson: {r} con p_value: {p_value}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)

print(f"Coeficiente de correlaci√≥n de Spearman: {rho}, valor p: {rho}")
```

+ Si analizamos el conjunto de 4 puntos de la parte superior derecaha tampoco hay correlaci√≥n. (ver ejemplos anteriores o calcularlos)
