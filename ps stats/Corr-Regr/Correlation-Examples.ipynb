````python
import numpy as np
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
````

+ Hip√≥tesis nula (ùêª0): No existe correlaci√≥n entre las variables ùëã e ùëå. 
    + Matem√°ticamente, esto se expresa como ùëü=0 para Pearson y ùúå=0 para Spearman.

+ Hip√≥tesis alternativa (ùêª1): Existe una correlaci√≥n significativa entre las variables ùëã e ùëå. 
    + Esto se expresa como ùëü‚â†0 para Searson o ùúå‚â†0 para Spearman.

#### Alta correlaci√≥n

````python
# Datos de ejemplo
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

# Grafico
plt.scatter(x, y, color='blue', label='Datos')

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)

print(f"Coeficiente de correlaci√≥n de Pearson: {r} con p_value: {p_value}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)

print(f"Coeficiente de correlaci√≥n de Spearman: {rho}, valor p: {rho}")
````

### 1. **Erupciones de volcanes**
+ Relaci√≥n entre la altura de la erupci√≥n de un volcan y los intervalos de tiempo posteriores a las erupciones del g√©iser Olf Fatihful.
    + Altura	           140	110	125	120	140	120	125	150
    + Intervalo posterior	92	65	72	94	83	94	101	87

+ ¬øExiste una correlaci√≥n lineal entre la altura de una erupci√≥n y el intervalo de tiempo posterior a la erupci√≥n?

````python
# Datos de ejemplo
x = [140, 110, 125, 120, 140, 120, 125, 150]  # Altura de la erupci√≥n
y = [92, 65, 72, 94, 83, 94, 101, 87]         # Intervalo de tiempo posterior a la erupci√≥n

# Gr√°fico de dispersi√≥n con l√≠nea de ajuste lineal
sns.regplot(x=x, y=y, color='red', line_kws={'color': 'blue'}, ci=None)

# Mostrar gr√°fico con etiquetas
plt.xlabel('Altura de la erupci√≥n')
plt.ylabel('Intervalo de tiempo posterior a la erupci√≥n')
plt.title('Gr√°fico de dispersi√≥n con l√≠nea de ajuste lineal')
plt.show()

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)
print(f"Coeficiente de correlaci√≥n de Pearson: {r:.2f}, valor p: {p_value:.2g}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)
print(f"Coeficiente de correlaci√≥n de Spearman: {rho:.2f}, valor p: {p_value:.2g}")
````

+ Por la grafica no se aprecia una correlaci√≥n significativa.
+ Por Pearson, se observa un indice de correlaci√≥n bajo con un p_value alto en favor de la H0 que los valores No est√°n dorrelacionados.
+ Por Spearman se observa un indice de correlaci√≥n muy bajo con un valor p que nos dice que su calculo puedo o no ser aplicable en este caso porque
    +  la relaci√≥n entre las variables no es mon√≥tona.
+ **Conclusi√≥n**: Alturas y tiempos, en este caso no est√°n correlacionados.

### 2. **Oyentes y ventas de canciones**
+ En la siguiente tabla se listan los n√∫meros de impresiones de oyentes (en cientos de millones) de canciones y los n√∫meros correspondientes de √°lbumes vendidos 
    + (en cientos de miles). El n√∫mero de impresiones de oyentes es un conteo del n√∫mero de veces que la gente ha escuchado la canci√≥n. 
    + La tabla se basa en datos de USA Today. ¬øParece que las ventas del √°lbum est√°n muy afectadas por el n√∫mero de impresiones de oyentes?
+ Impresiones	de oyentes	28	13	14	24	20	18	14	24	17
+ √Ålbumes vendidos	    19	7	7	20	6	4	5	25	12

````python
# Datos de ejemplo
x = [28, 13, 14, 24, 20, 18, 14, 24, 17]  # Impresiones de canci√≥n
y = [19, 7, 7, 20, 6, 4, 5, 25, 12]       # Albumes vendidos

# Gr√°fico de dispersi√≥n con l√≠nea de ajuste lineal
sns.regplot(x=x, y=y, color='red', line_kws={'color': 'blue'}, ci=None)

# Mostrar gr√°fico con etiquetas
plt.xlabel('Impresiones (MM)')
plt.ylabel('Albumes (M)')
plt.title('Dispersi√≥n con l√≠nea de ajuste lineal')
plt.show()

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)
print(f"Coeficiente de correlaci√≥n de Pearson: {r:.2f}, valor p: {p_value:.2g}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)
print(f"Coeficiente de correlaci√≥n de Spearman: {rho:.2f}, valor p: {p_value:.2g}")
````

**Coeficiente de Correlaci√≥n de Pearson**
    + El coeficiente de correlaci√≥n de Pearson mide la fuerza y direcci√≥n de una relaci√≥n lineal entre dos variables. El valor calculado fue ùëü = 0.80 con un valor p de 0.01.
**Interpretaci√≥n**
+ r=0.80 indica una correlaci√≥n positiva fuerte. A medida que aumentan las impresiones de oyentes, las ventas de √°lbumes tienden a aumentar.
+ Un valor p de 0.01 es muy peque√±o, lo que sugiere que la correlaci√≥n es estad√≠sticamente significativa. 
+ Esto implica que es poco probable que la relaci√≥n observada entre las impresiones de oyentes y las ventas de √°lbumes se deba al azar.

**Coeficiente de Correlaci√≥n de Spearman**
+ El coeficiente de correlaci√≥n de Spearman mide la fuerza y direcci√≥n de la relaci√≥n mon√≥tona (no necesariamente lineal) entre dos variables. 
+ El valor calculado fue  œÅ=0.56 con un valor p de 0.12.

**Interpretaci√≥n**
+ œÅ=0.56 indica una correlaci√≥n positiva moderada, pero no tan fuerte como la relaci√≥n lineal.
+ Un valor p de 0.12 no es suficientemente peque√±o (generalmente necesitamos ùëù < 0.05 para significancia), 
    + lo que sugiere que la correlaci√≥n no es estad√≠sticamente significativa. 
+ Esto significa que no podemos descartar la posibilidad de que la relaci√≥n observada se deba al azar.

**Interpretaci√≥n y Conclusi√≥n**
**Diferencia entre Pearson y Spearman**
+ Pearson mide la correlaci√≥n lineal, mientras que Spearman mide la correlaci√≥n mon√≥tona. 
+ La diferencia en los coeficientes sugiere que, aunque existe una relaci√≥n lineal fuerte entre las impresiones de oyentes y las ventas de √°lbumes, 
    + la relaci√≥n puede no ser perfectamente mon√≥tona.
+ La diferencia en los valores p tambi√©n indica que la relaci√≥n lineal es m√°s robusta (estad√≠sticamente significativa) que una relaci√≥n mon√≥tona general.

**Conclusi√≥n**
+ Pearson: La correlaci√≥n lineal fuerte y significativa sugiere que el n√∫mero de impresiones de oyentes tiene un efecto considerable en las ventas de √°lbumes. 
    + Esto indica que a medida que m√°s personas escuchan una canci√≥n, es m√°s probable que se vendan m√°s √°lbumes.
+ Spearman: La correlaci√≥n no es tan fuerte ni significativa, lo que indica que la relaci√≥n entre las variables no sigue necesariamente un patr√≥n mon√≥tono claro, 
    + aunque hay una tendencia general.
+ En resumen, el an√°lisis sugiere que las ventas de √°lbumes est√°n significativamente afectadas por el n√∫mero de impresiones de oyentes, 
    + especialmente si la relaci√≥n es lineal. 
    + La relaci√≥n mon√≥tona, sin embargo, es menos clara y menos significativa.

### 3. **Presupuestos e ingresos brutos de pel√≠culas**
+ En la siguiente tabla se muestran los presupuestos (en millones de d√≥lares) y los ingresos brutos (en millones de d√≥lares) 
    + de pel√≠culas seleccionadas al azar (seg√∫n datos de la Motion Picture Association of America). 
    + Presupuesto	62	90	50	35	200	100	90
    + Ingresos	65	64	48	57	601	146	47				
+ ¬øParece existir una correlaci√≥n lineal entre el dinero gastado para filmar la pel√≠cula y la cantidad recuperada en las salas de cine? 
    + Adem√°s del monto del presupuesto, identifique otro factor importante que puede afectar la cantidad de dinero que obtiene la pel√≠cula.

````python
# Datos de ejemplo
x = [62, 90, 50, 35, 200, 100, 90]  # Presupuesto
y = [65, 64, 48, 57, 601, 146, 47]  # Ingresos

# Gr√°fico de dispersi√≥n con l√≠nea de ajuste lineal
sns.regplot(x=x, y=y, color='red', line_kws={'color': 'blue'}, ci=None)

# Mostrar gr√°fico con etiquetas
plt.xlabel('Presupuesto (MM)')
plt.ylabel('Ingresos (MM)')
plt.title('Dispersi√≥n con l√≠nea de ajuste lineal')
plt.show()

# C√°lculo del coeficiente de correlaci√≥n de Pearson
r, p_value = pearsonr(x, y)
print(f"Coeficiente de correlaci√≥n de Pearson: {r:.2f}, valor p: {p_value:.2g}")

# C√°lculo del coeficiente de correlaci√≥n de Spearman
rho, p_value = spearmanr(x, y)
print(f"Coeficiente de correlaci√≥n de Spearman: {rho:.2f}, valor p: {p_value:.2g}")
````

+ Se observa una muy alta correlaci√≥n entre ambas variables. Esta correlaci√≥n est√° apoyada por:
    + Un alto √≠ndice de correlaic√≥n de Pearson.
    + Un bajo valor p < 0.01 que desestima la posibilidad de la H0 que ambas variables no est√©n correlacionadas.
+ El coeficiente de correlaci√≥n de Spearman arroja evidencia (observada) que la correlaci√≥n existe (> 0.5 pero no es mon√≥tona)
**Conclusi√≥n**: 
+ Existe correlaci√≥n entre el presupuesto y los ingresos que la pel√≠cula genera.
+ Existen tambi√©n otros factores que inciden en los ingresos que genra la pelicula:
    + Las criticas del avant premiere
    + La calidad de los actores
    + El g√©nero y la trama
