# Regresiones Polinómicas
### Ejemplos de ajustes polinómicos
###### https://www.chreinvent.com/recursos/regresi%C3%B3n-no-lineal

**Primer ejemplo**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, root_mean_squared_error

# Generar datos simulados
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # Variable independiente
y = 2.5 * X**2 + np.random.randn(100, 1) * 5  # Variable dependiente con ruido

# Transformación polinómica
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Ajuste del modelo de regresión polinómica
modelo_polinomico = LinearRegression()
modelo_polinomico.fit(X_poly, y)

# Predicciones
X_new = np.linspace(0, 10, 100).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
y_pred = modelo_polinomico.predict(X_new_poly)

# Gráfico
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X_new, y_pred, color='red', label='Ajuste polinómico (grado 2)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Regresión Polinómica')
plt.show()

# Parámetros obtenidos
coeficientes = modelo_polinomico.coef_
intercepto = modelo_polinomico.intercept_
print(f"Coeficientes: {coeficientes}")
print(f"Intercepto: {intercepto}")

# Cálculo de MSE y RMSE
mse =mean_squared_error(y, y_pred)
print(f"MSE (R2): {mse}")

rmse =root_mean_squared_error(y, y_pred)
print(f"RMSE: {rmse}")

# Calcular y mostrar R^2
r2_score = modelo_polinomico.score(X_poly, y)
print(f"Coeficiente de determinación R^2: {r2_score}")
```

**Segundo Ejemplo**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error

# Datos simulados
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Valores entre 0 y 10
y = 2 + 3*X - 0.5*X**2 + np.random.randn(100, 1)*5  # Relación cuadrática con ruido

# Transformación polinómica (grado 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Ajustar modelo
model = LinearRegression()
model.fit(X_poly, y)

# Predicciones
X_new = np.linspace(0, 10, 100).reshape(-1, 1)
X_new_poly = poly.transform(X_new)
y_pred = model.predict(X_new_poly)

# Gráfico
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X_new, y_pred, color='red', label='Ajuste polinómico (grado 2)')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Regresión Polinómica')
plt.show()

# Parámetros obtenidos
coeficientes = model.coef_
intercepto = model.intercept_
print(f"Coeficientes: {coeficientes}")
print(f"Intercepto: {intercepto}")

# Cálculo de MSE y RMSE
mse =mean_squared_error(y, y_pred)
print(f"MSE (R2): {mse}")

rmse =root_mean_squared_error(y, y_pred)
print(f"RMSE: {rmse}")

# Calcular y mostrar R^2
r2_score = model.score(X_poly, y)
print(f"Coeficiente de determinación R^2: {r2_score}")
```
