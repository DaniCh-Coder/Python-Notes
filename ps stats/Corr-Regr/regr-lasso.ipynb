```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split

# Generando datos simulados
np.random.seed(0)
X = np.random.rand(100, 5)  # 5 variables independientes
y = 3*X[:,0] + 1.5*X[:,1] + np.random.randn(100) * 0.5  # Variable dependiente con ruido

# División de los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Ajuste del modelo de regresión Lasso
modelo_lasso = Lasso(alpha=0.1)  # λ = 0.1
modelo_lasso.fit(X_train, y_train)
y_pred = modelo_lasso.predict(X_test)

# Resultados
print("Coeficientes de la Regresión Lasso:", modelo_lasso.coef_)
print("Intersección (β0):", modelo_lasso.intercept_)

# Gráfico de los resultados
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, label='Predicciones')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', label='Línea Ideal')
plt.xlabel('Valores Reales')
plt.ylabel('Valores Predichos')
plt.title('Regresión Lasso: Valores Reales vs Predichos')
plt.legend()
plt.show()
```
