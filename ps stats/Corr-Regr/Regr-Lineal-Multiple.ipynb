# Regresión Lineal Múltiple

+ Queremos predecir el precio de una casa (variable dependiente) usando tres factores (variables independientes): 
    + tamaño de la casa (en metros cuadrados), número de habitaciones y edad de la casa.
+ La ecuación de la regresión múltiple sería: Precio=β0+β1⋅Tamaño+β2⋅Habitaciones+β3⋅Edad+ε
    + donde:
        + β1 nos dice cuánto cambia el precio de la casa por cada metro cuadrado adicional, manteniendo constantes las otras dos variables.
        + β2 nos dice cómo cambia el precio con cada habitación adicional.
        + β3 nos dice cómo cambia el precio con el aumento en la edad de la casa.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Datos de ejemplo
data = {
    'Tamaño': [1500, 1800, 2400, 3000, 3500],
    'Habitaciones': [3, 3, 4, 4, 5],
    'Edad': [10, 15, 8, 5, 20],
    'Precio': [300000, 350000, 400000, 500000, 550000]
}

# Convertimos los datos en un DataFrame
df = pd.DataFrame(data)

# Variables independientes (predictoras)
X = df[['Tamaño', 'Habitaciones', 'Edad']]

# Variable dependiente (respuesta)
y = df['Precio']

# Añadimos una constante al modelo (intercepto)
X = sm.add_constant(X)

# Ajustamos el modelo de regresión múltiple
modelo = sm.OLS(y, X).fit()

# Mostramos el resumen del modelo
print(modelo.summary())
```

```python
import matplotlib.pyplot as plt

# Predicciones basadas en el modelo ajustado
df['Predicciones'] = modelo.predict(X)

# Gráfico de los valores observados vs predicciones
plt.figure(figsize=(10,6))

# Gráfica de puntos
plt.scatter(df['Precio'], df['Predicciones'], color='blue', label='Datos')

# Línea de regresión perfecta (si predicción fuera igual a valor real)
plt.plot([df['Precio'].min(), df['Precio'].max()], [df['Precio'].min(), df['Precio'].max()], color='red', linestyle='--', label='Regresión Ideal')

# Título y etiquetas
plt.title('Regresión Múltiple: Valores Observados vs. Predichos', fontsize=16)
plt.xlabel('Valores Observados (Precio Real)', fontsize=12)
plt.ylabel('Valores Predichos (Precio Estimado)', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()

```
