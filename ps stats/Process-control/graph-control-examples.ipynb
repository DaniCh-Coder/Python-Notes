#### Control de Medias

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
np.random.seed(0)
data = np.random.normal(loc=10, scale=1, size=30)

# Calcular la media y los l√≠mites de control
mean = np.mean(data)
UCL = mean + 3 * np.std(data)
LCL = mean - 3 * np.std(data)

# Crear el gr√°fico de control
plt.figure(figsize=(10, 6))
plt.plot(data, marker='o', linestyle='-', color='blue', label='Datos')
plt.axhline(mean, color='green', linestyle='--', label='Media')
plt.axhline(UCL, color='red', linestyle='--', label='L√≠mite Superior de Control (UCL)')
plt.axhline(LCL, color='red', linestyle='--', label='L√≠mite Inferior de Control (LCL)')
plt.title('Gr√°fico de Control de Medias (X-barra)')
plt.xlabel('Muestra')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

#### Control de Rangos

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
np.random.seed(0)
data = np.random.normal(loc=10, scale=1, size=(10, 5))  # 10 muestras de tama√±o 5

# Calcular los rangos
rangos = np.ptp(data, axis=1)

# Calcular la media y los l√≠mites de control
mean_r = np.mean(rangos)
UCL_r = mean_r + 3 * np.std(rangos)
LCL_r = mean_r - 3 * np.std(rangos)

# Crear el gr√°fico de control de rango
plt.figure(figsize=(10, 6))
plt.plot(rangos, marker='o', linestyle='-', color='blue', label='Rangos')
plt.axhline(mean_r, color='green', linestyle='--', label='Media')
plt.axhline(UCL_r, color='red', linestyle='--', label='L√≠mite Superior de Control (UCL)')
plt.axhline(LCL_r, color='red', linestyle='--', label='L√≠mite Inferior de Control (LCL)')
plt.title('Gr√°fico de Control de Rango (R)')
plt.xlabel('Muestra')
plt.ylabel('Rango')
plt.legend()
plt.show()
```

#### Control de proporciones

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
np.random.seed(0)
n = 50  # Tama√±o de la muestra
defectuosos = np.random.binomial(n, 0.1, size=20)  # 20 muestras

# Calcular las proporciones
proporciones = defectuosos / n

# Calcular la media y los l√≠mites de control
mean_p = np.mean(proporciones)
UCL_p = mean_p + 3 * np.sqrt((mean_p * (1 - mean_p)) / n)
LCL_p = mean_p - 3 * np.sqrt((mean_p * (1 - mean_p)) / n)

# Crear el gr√°fico de control de proporciones
plt.figure(figsize=(10, 6))
plt.plot(proporciones, marker='o', linestyle='-', color='blue', label='Proporciones')
plt.axhline(mean_p, color='green', linestyle='--', label='Media')
plt.axhline(UCL_p, color='red', linestyle='--', label='L√≠mite Superior de Control (UCL)')
plt.axhline(LCL_p, color='red', linestyle='--', label='L√≠mite Inferior de Control (LCL)')
plt.title('Gr√°fico de Control de Proporciones (p)')
plt.xlabel('Muestra')
plt.ylabel('Proporci√≥n de Defectuosos')
plt.legend()
plt.show()
```

#### Control de defectos

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos simulados
np.random.seed(0)
defectos = np.random.poisson(lam=5, size=20)  # 20 muestras

# Calcular la media y los l√≠mites de control
mean_c = np.mean(defectos)
UCL_c = mean_c + 3 * np.sqrt(mean_c)
LCL_c = mean_c - 3 * np.sqrt(mean_c)

# Crear el gr√°fico de control de n√∫mero de defectos
plt.figure(figsize=(10, 6))
plt.plot(defectos, marker='o', linestyle='-', color='blue', label='Defectos')
plt.axhline(mean_c, color='green', linestyle='--', label='Media')
plt.axhline(UCL_c, color='red', linestyle='--', label='L√≠mite Superior de Control (UCL)')
plt.axhline(LCL_c, color='red', linestyle='--', label='L√≠mite Inferior de Control (LCL)')
plt.title('Gr√°fico de Control de N√∫mero de Defectos (c)')
plt.xlabel('Muestra')
plt.ylabel('N√∫mero de Defectos')
plt.legend()
plt.show()
```

#### Control de rachas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos simulados de ventas
np.random.seed(0)
ventas = np.random.randint(50, 150, size=30)

# Calcular el promedio de ventas
promedio_ventas = np.mean(ventas)

# Clasificar los d√≠as como "buenos" (1) o "malos" (0)
dias = np.where(ventas > promedio_ventas, 1, 0)

# Contar el n√∫mero de rachas
def contar_rachas(dias):
    rachas = 1
    for i in range(1, len(dias)):
        if dias[i] != dias[i-1]:
            rachas += 1
    return rachas

# Calcular el n√∫mero de rachas observadas
rachas_observadas = contar_rachas(dias)

# Calcular el n√∫mero esperado de rachas y su desviaci√≥n est√°ndar
n1 = np.sum(dias)
n2 = len(dias) - n1
rachas_esperadas = ((2 * n1 * n2) / (n1 + n2)) + 1
desviacion_estandar_rachas = np.sqrt((2 * n1 * n2 * (2 * n1 * n2 - n1 - n2)) / ((n1 + n2)**2 * (n1 + n2 - 1)))

# Calcular el valor Z
z = (rachas_observadas - rachas_esperadas) / desviacion_estandar_rachas

# Determinar el valor cr√≠tico para un nivel de significancia del 5%
valor_critico = norm.ppf(1 - 0.05 / 2)

# Gr√°fico de las ventas
plt.figure(figsize=(10, 6))
plt.plot(ventas, marker='o', linestyle='-', color='blue', label='Ventas')
plt.axhline(promedio_ventas, color='green', linestyle='--', label='Promedio de Ventas')
plt.title('Ventas Diarias')
plt.xlabel('D√≠a')
plt.ylabel('Ventas')
plt.legend()
plt.show()

# Resultados
print(f"Ventas: {ventas}")
print(f"D√≠as (1 = Bueno, 0 = Malo): {dias}")
print(f"Rachas Observadas: {rachas_observadas}")
print(f"Rachas Esperadas: {rachas_esperadas:.2f}")
print(f"Desviaci√≥n Est√°ndar de Rachas: {desviacion_estandar_rachas:.2f}")
print(f"Valor Z: {z:.2f}")
print(f"Valor Cr√≠tico (5%): {valor_critico:.2f}")

# Conclusi√≥n
if abs(z) > valor_critico:
    print("Rechazamos la hip√≥tesis nula. Hay evidencia de que las ventas no ocurren de manera aleatoria.")
else:
    print("No rechazamos la hip√≥tesis nula. No hay evidencia de que las ventas no ocurran de manera aleatoria.")
```
