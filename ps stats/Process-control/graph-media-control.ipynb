# Grafico de control de media

Ejemplo: Control peso de botellas
+ Datos Simulados: Supongamos que tienes 30 valores de peso promedio de botellas en gramos. 
+ Cálculos:Media general de los pesos.

Límite de control superior (UCL) y límite de control inferior (LCL), basados en la desviación estándar de las medias.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo de Seaborn
sns.set(style="whitegrid")

# Datos simulados: peso promedio de 5 botellas por día durante 30 días
pesos_promedio = [
    501.2, 499.8, 500.5, 502.3, 498.6, 499.4, 501.8, 500.9, 500.0, 501.1,
    498.9, 500.4, 502.5, 499.2, 500.8, 500.3, 499.7, 501.0, 500.2, 501.4,
    499.5, 500.6, 502.0, 498.8, 500.7, 501.3, 500.1, 499.3, 501.9, 500.5
]

# Cálculos
media_general = np.mean(pesos_promedio)
desviacion_estandar = np.std(pesos_promedio, ddof=1) / np.sqrt(len(pesos_promedio))
UCL = media_general + 3 * desviacion_estandar
LCL = media_general - 3 * desviacion_estandar

# Crear el gráfico usando Seaborn y Matplotlib
plt.figure(figsize=(12, 6))
sns.lineplot(x=range(1, len(pesos_promedio) + 1), y=pesos_promedio, marker='o', color='b', label='Peso promedio')
plt.axhline(y=media_general, color='green', linestyle='--', label='Media General')
plt.axhline(y=UCL, color='red', linestyle='--', label='Límite de Control Superior (UCL)')
plt.axhline(y=LCL, color='red', linestyle='--', label='Límite de Control Inferior (LCL)')
plt.title('Gráfico de Control de Media (\u0304X) - Peso Promedio de Botellas', fontsize=14)
plt.xlabel('Día', fontsize=12)
plt.ylabel('Peso Promedio (g)', fontsize=12)
plt.legend()
plt.show()
```
