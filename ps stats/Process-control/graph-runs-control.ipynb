# Control de rachas

```python
import matplotlib.pyplot as plt

# Secuencia de datos: N para no defectuoso, D para defectuoso
secuencia = ['N', 'N', 'D', 'D', 'N', 'D', 'N', 'N', 'D', 'D', 'D', 'N', 'D', 'N', 'N']

# Convertir la secuencia en binario: N -> 0, D -> 1
secuencia_binaria = [0 if x == 'N' else 1 for x in secuencia]

# Contar el número de rachas
def contar_rachas(secuencia):
    rachas = 1  # Siempre hay al menos una racha
    for i in range(1, len(secuencia)):
        if secuencia[i] != secuencia[i - 1]:
            rachas += 1
    return rachas

num_rachas = contar_rachas(secuencia_binaria)

print("Número de rachas:", num_rachas)
```

```python
# Gráfico de la secuencia con las rachas
plt.figure(figsize=(10, 4))
plt.plot(secuencia_binaria, marker='o', linestyle='-', color='b', label='Secuencia Binaria')
plt.title('Gráfico de Control de Rachas')
plt.xlabel('Índice')
plt.ylabel('Categoría (0 = No Defectuoso, 1 = Defectuoso)')
plt.yticks([0, 1], ['No Defectuoso', 'Defectuoso'])
plt.axhline(y=0.5, color='grey', linestyle='--', alpha=0.5)  # Línea de referencia
plt.legend()
plt.show()
```
