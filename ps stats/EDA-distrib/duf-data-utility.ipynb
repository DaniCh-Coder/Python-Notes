# Funci√≥n de utilidad de los datos
###### https://www.chreinvent.com/recursos/datos-para-un-proyecto

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Crear variables independientes
calidad = np.linspace(0, 1, 100)  # Calidad de los datos (0 a 1)
anonimizacion = np.linspace(0, 1, 100)  # Nivel de anonimizaci√≥n (0 a 1)

# Crear una malla para los ejes
calidad_mesh, anonimizacion_mesh = np.meshgrid(calidad, anonimizacion)

# Definir una funci√≥n de utilidad arbitraria
DUF = calidad_mesh * (1 - anonimizacion_mesh) ** 2

# Crear el gr√°fico 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(calidad_mesh, anonimizacion_mesh, DUF, cmap='viridis', edgecolor='k')

# Etiquetas y t√≠tulo
ax.set_xlabel('Calidad de los datos')
ax.set_ylabel('Nivel de anonimizaci√≥n')
ax.set_zlabel('Utilidad de los datos (DUF)')
ax.set_title('Funci√≥n de utilidad de los datos (DUF)')

# Barra de colores
fig.colorbar(surface, shrink=0.5, aspect=10)

plt.show()
```
