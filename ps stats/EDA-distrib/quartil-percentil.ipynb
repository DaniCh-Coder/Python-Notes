# Cuartiles y Percentiles

1. Distribuci√≥n normal con cuartiles y percentiles

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generar datos de una distribuci√≥n normal
np.random.seed(0)
mu, sigma = 0, 1  # media y desviaci√≥n est√°ndar
datos = np.random.normal(mu, sigma, 1000)

# Calcular los cuartiles
Q1 = np.percentile(datos, 25)
Q2 = np.percentile(datos, 50)
Q3 = np.percentile(datos, 75)

# Calcular percentiles espec√≠ficos (por ejemplo, P30 y P90)
P30 = np.percentile(datos, 30)
P90 = np.percentile(datos, 90)

# Crear gr√°fico
fig, ax = plt.subplots(figsize=(10, 6))

# Histograma de la distribuci√≥n normal
ax.hist(datos, bins=30, density=True, alpha=0.6, color='skyblue')

# Curva de densidad
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, sigma)
ax.plot(x, p, 'k', linewidth=2)

# Marcar cuartiles en el gr√°fico
for quartile, color, label in zip([Q1, Q2, Q3], ['red', 'green', 'red'], ['Q1', 'Q2 (Mediana)', 'Q3']):
    ax.axvline(quartile, color=color, linestyle='--', label=f'{label} = {quartile:.2f}')

# Marcar percentiles en el gr√°fico
for percentile, color, label in zip([P30, P90], ['purple', 'orange'], ['P30', 'P90']):
    ax.axvline(percentile, color=color, linestyle='--', label=f'{label} = {percentile:.2f}')

# Configuraci√≥n del gr√°fico
ax.set_title('Distribuci√≥n Normal con Cuartiles y Percentiles')
ax.set_xlabel('Valor')
ax.set_ylabel('Densidad de Probabilidad')
ax.legend()
ax.grid(True)

# Mostrar gr√°fico
plt.show()
```

2. Boxplot de la distribuci√≥n normal

```python
import seaborn as sns

# Generar datos de una distribuci√≥n normal
np.random.seed(0)
mu, sigma = 0, 1  # media y desviaci√≥n est√°ndar
datos_normal = np.random.normal(mu, sigma, 1000)

# Crear gr√°fico de boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=datos_normal, orient='h', color='skyblue')
plt.title('Boxplot de una Distribuci√≥n Normal')
plt.xlabel('Valor')
plt.grid(True)
plt.show()
```

3. Distribuci√≥n normal y boxplot con cuartiles

```python

# Generar datos de una distribuci√≥n normal
np.random.seed(0)
mu, sigma = 0, 1  # media y desviaci√≥n est√°ndar
datos_normal = np.random.normal(mu, sigma, 1000)

# Crear figura y subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# Primer subplot: Histograma de la distribuci√≥n normal
axs[0].hist(datos_normal, bins=30, density=True, alpha=0.6, color='skyblue')
xmin, xmax = axs[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu, sigma)
axs[0].plot(x, p, 'k', linewidth=2)
axs[0].set_title('Distribuci√≥n Normal')
axs[0].set_xlabel('Valor')
axs[0].set_ylabel('Densidad de Probabilidad')
axs[0].grid(True)

# Segundo subplot: Boxplot de la distribuci√≥n normal
sns.boxplot(data=datos_normal, orient='h', ax=axs[1], color='skyblue')
axs[1].set_title('Boxplot de la Distribuci√≥n Normal')
axs[1].set_xlabel('Valor')
axs[1].grid(True)

# Ajustar el dise√±o para que no se superpongan los subplots
plt.tight_layout()
plt.show()
```

4. Distribuci√≥n Exponencial con cuartiles y percentiles

```python
# Generar datos de una distribuci√≥n exponencial
np.random.seed(0)
scale = 1  # par√°metro de escala (1/lambda)
datos = np.random.exponential(scale, 1000)

# Calcular los cuartiles
Q1 = np.percentile(datos, 25)
Q2 = np.percentile(datos, 50)
Q3 = np.percentile(datos, 75)

# Calcular percentiles espec√≠ficos (por ejemplo, P30 y P90)
P30 = np.percentile(datos, 30)
P90 = np.percentile(datos, 90)

# Crear gr√°fico
fig, ax = plt.subplots(figsize=(10, 6))

# Histograma de la distribuci√≥n exponencial
ax.hist(datos, bins=30, density=True, alpha=0.6, color='skyblue')

# Curva de densidad
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.expon.pdf(x, scale=scale)
ax.plot(x, p, 'k', linewidth=2)

# Marcar cuartiles en el gr√°fico
for quartile, color, label in zip([Q1, Q2, Q3], ['red', 'green', 'red'], ['Q1', 'Q2 (Mediana)', 'Q3']):
    ax.axvline(quartile, color=color, linestyle='--', label=f'{label} = {quartile:.2f}')

# Marcar percentiles en el gr√°fico
for percentile, color, label in zip([P30, P90], ['purple', 'orange'], ['P30', 'P90']):
    ax.axvline(percentile, color=color, linestyle='--', label=f'{label} = {percentile:.2f}')

# Configuraci√≥n del gr√°fico
ax.set_title('Distribuci√≥n Exponencial con Cuartiles y Percentiles')
ax.set_xlabel('Valor')
ax.set_ylabel('Densidad de Probabilidad')
ax.legend()
ax.grid(True)

# Mostrar gr√°fico
plt.show()
```
