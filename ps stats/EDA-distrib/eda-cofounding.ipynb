# Confusión - Cofounding
La confusión experimental, también conocida como "confounding," ocurre cuando el efecto de una variable independiente sobre una variable dependiente está distorsionado por la presencia de otra variable que está relacionada con ambas. Esta variable adicional se conoce como "variable de confusión."

Supongamos que estamos estudiando la relación entre el consumo de helado (Variable A) y la tasa de ahogamientos (Variable B). 
+ Observamos que a medida que aumenta el consumo de helado, también aumenta la tasa de ahogamientos. 
+ Sin embargo, podría haber una variable de confusión, como la temperatura (Variable C), que influye tanto en el consumo de helado como en la tasa de ahogamientos.

Ejecutando este código, se ve cómo la variable de confusión (temperatura) puede influir en la relación aparente entre dos variables, 
+ llevando a conclusiones erróneas si no se tiene en cuenta.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
np.random.seed(0)
temperatura = np.random.uniform(20, 40, 100)  # Temperatura en grados Celsius
consumo_helado = 2 * temperatura + np.random.normal(0, 5, 100)  # Consumo de helado en litros
tasa_ahogamientos = 0.5 * temperatura + np.random.normal(0, 2, 100)  # Tasa de ahogamientos por 100,000 habitantes

# Gráficos
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Gráfico de consumo de helado vs. tasa de ahogamientos (aparente)
axs[0].scatter(consumo_helado, tasa_ahogamientos, color='blue')
axs[0].set_title("Consumo de Helado vs. Tasa de Ahogamientos")
axs[0].set_xlabel("Consumo de Helado (litros)")
axs[0].set_ylabel("Tasa de Ahogamientos")

# Gráfico de temperatura vs. consumo de helado (variable de confusión)
axs[1].scatter(temperatura, consumo_helado, color='green')
axs[1].set_title("Temperatura vs. Consumo de Helado")
axs[1].set_xlabel("Temperatura (°C)")
axs[1].set_ylabel("Consumo de Helado (litros)")

# Gráfico de temperatura vs. tasa de ahogamientos (variable de confusión)
axs[2].scatter(temperatura, tasa_ahogamientos, color='red')
axs[2].set_title("Temperatura vs. Tasa de Ahogamientos")
axs[2].set_xlabel("Temperatura (°C)")
axs[2].set_ylabel("Tasa de Ahogamientos")

# Ajustar diseño
plt.tight_layout()
plt.show()
```
