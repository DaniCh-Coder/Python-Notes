# Ejemplo para poblaciones infitinas o grandes
Supongamos que un investigador quiere estimar la media del colesterol total en la sangre de los adultos de una ciudad. 
Toma una muestra aleatoria de 100 adultos y encuentra una media muestral (Xˉ) de 200 mg/dL con una desviación estándar muestral (s) de 15 mg/dL. Desea construir un intervalo de confianza del 95% para la media del colesterol total.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de la muestra
n = 100
mean_sample = 200
std_sample = 15
confidence_level = 0.95

# Calcular el valor crítico (t de Student) para el nivel de confianza dado
alpha = 1 - confidence_level
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)

# Calcular el margen de error
margin_of_error = t_critical * (std_sample / np.sqrt(n))

# Calcular el intervalo de confianza
lower_bound = mean_sample - margin_of_error
upper_bound = mean_sample + margin_of_error

print(f'Intervalo de Confianza del 95%: ({lower_bound:.2f}, {upper_bound:.2f})')

# Representación gráfica mejorada
fig, ax = plt.subplots(figsize=(12, 6))

# Histograma de la muestra original (simulada)
data = np.random.normal(mean_sample, std_sample, n)
ax.hist(data, bins=20, alpha=0.6, color='lightblue', edgecolor='black', density=True, label='Datos de la muestra')

# Curva de distribución normal
x = np.linspace(mean_sample - 4*std_sample, mean_sample + 4*std_sample, 1000)
y = stats.norm.pdf(x, mean_sample, std_sample)
ax.plot(x, y, 'b-', lw=2, label='Distribución Normal')

# Intervalo de confianza
ax.axvline(lower_bound, color='r', linestyle='--', label=f'Límite Inferior del IC = {lower_bound:.2f}')
ax.axvline(upper_bound, color='r', linestyle='--', label=f'Límite Superior del IC = {upper_bound:.2f}')
ax.axvline(mean_sample, color='g', linestyle='-', label=f'Media Muestral = {mean_sample}')

# Etiquetas y título
ax.set_xlabel('Colesterol Total (mg/dL)')
ax.set_ylabel('Densidad')
ax.set_title('Intervalo de Confianza del 95% para la Media del Colesterol Total')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**
Intervalo de Confianza del 95%: 
+ Suponiendo que los cálculos sean correctos, el intervalo de confianza del 95% para la media del colesterol total sería de aproximadamente (197.08, 202.92) mg/dL.

Significado: Esto significa que estamos 95% seguros de que la verdadera media del colesterol total en la población está entre 197.08 y 202.92 mg/dL.

## Regla del intervalo

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución normal
mu = 0  # media
sigma = 1  # desviación estándar

# Generar datos de una distribución normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Crear gráfico
plt.figure(figsize=(12, 6))

# Graficar la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombrear el área de +/- 1 desviación estándar (68% de los datos)
plt.fill_between(x, y, where=((x >= mu - sigma) & (x <= mu + sigma)), color='skyblue', alpha=0.5, label='68% (1σ)')

# Sombrear el área de +/- 2 desviaciones estándar (95% de los datos)
plt.fill_between(x, y, where=((x >= mu - 2*sigma) & (x <= mu + 2*sigma)), color='lightgreen', alpha=0.5, label='95% (2σ)')

# Sombrear el área de +/- 3 desviaciones estándar (99.7% de los datos)
plt.fill_between(x, y, where=((x >= mu - 3*sigma) & (x <= mu + 3*sigma)), color='salmon', alpha=0.5, label='99.7% (3σ)')

# Configuración del gráfico
plt.title('Ejemplo de la Regla del Intervalo (Regla 68-95-99.7)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

## Ejemplo para poblaciones finitas
Supongamos que un investigador quiere estimar la media de las puntuaciones de un examen de una escuela con 500 estudiantes. 
+ El investigador toma una muestra aleatoria de 50 estudiantes y encuentra una media muestral (Xˉ) de 70 puntos con una desviación estándar muestral (σ) de 10 puntos. 
+ Desea construir un intervalo de confianza del 95% para la media de las puntuaciones del examen.

```python
import numpy as np
import scipy.stats as stats

# Datos de la muestra y población
N = 500  # Tamaño de la población
n = 50  # Tamaño de la muestra
mean_sample = 70
std_sample = 10
confidence_level = 0.95

# Calcular el valor crítico (Z) para el nivel de confianza dado
alpha = 1 - confidence_level
Z_critical = stats.norm.ppf(1 - alpha/2)

# Factor de Corrección de Población Finita (FPC)
FPC = np.sqrt((N - n) / (N - 1))

# Calcular el margen de error
margin_of_error = Z_critical * (std_sample / np.sqrt(n)) * FPC

# Calcular el intervalo de confianza
lower_bound = mean_sample - margin_of_error
upper_bound = mean_sample + margin_of_error

print(f'Intervalo de Confianza del 95%: ({lower_bound:.2f}, {upper_bound:.2f})')

# Representación gráfica del intervalo de confianza
import matplotlib.pyplot as plt

# Datos para el gráfico
x = np.linspace(mean_sample - 4*std_sample, mean_sample + 4*std_sample, 1000)
y = stats.norm.pdf(x, mean_sample, std_sample)

# Crear el gráfico
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(x, y, 'b-', lw=2, label='Distribución Normal')
ax.axvline(lower_bound, color='r', linestyle='--', label=f'Límite Inferior del IC = {lower_bound:.2f}')
ax.axvline(upper_bound, color='r', linestyle='--', label=f'Límite Superior del IC = {upper_bound:.2f}')
ax.axvline(mean_sample, color='g', linestyle='-', label=f'Media Muestral = {mean_sample}')

# Etiquetas y título
ax.set_xlabel('Puntuación del Examen')
ax.set_ylabel('Densidad')
ax.set_title('Intervalo de Confianza del 95% para la Media de las Puntuaciones del Examen (Población Finita)')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Intervalo de Confianza del 95%:
+ Según el cálculo, el intervalo de confianza del 95% para la media de las puntuaciones del examen en la población es aproximadamente (67.26, 72.74) puntos.
+ Esto significa que estamos 95% seguros de que la verdadera media de las puntuaciones del examen en la población de 500 estudiantes está entre 67.26 y 72.74 puntos.

Representación Gráfica:
+ La curva azul representa la distribución normal ajustada a los datos de la muestra.
+ Las líneas rojas discontinuas indican los límites inferior y superior del intervalo de confianza.
+ La línea verde representa la media muestral.

Este ejemplo ilustra cómo ajustar los cálculos de intervalo de confianza para una población finita, proporcionando una estimación más precisa del parámetro poblacional.
