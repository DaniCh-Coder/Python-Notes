# Intervalo de confianza
## Regla del intervalo

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución normal
mu = 0  # media
sigma = 1  # desviación estándar

# Generar datos de una distribución normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Crear gráfico
plt.figure(figsize=(12, 6))

# Graficar la distribución normal
plt.plot(x, y, label='Distribución Normal')

# Sombrear el área de +/- 1 desviación estándar (68% de los datos)
plt.fill_between(x, y, where=((x >= mu - sigma) & (x <= mu + sigma)), color='skyblue', alpha=0.5, label='68% (1σ)')

# Sombrear el área de +/- 2 desviaciones estándar (95% de los datos)
plt.fill_between(x, y, where=((x >= mu - 2*sigma) & (x <= mu + 2*sigma)), color='lightgreen', alpha=0.5, label='95% (2σ)')

# Sombrear el área de +/- 3 desviaciones estándar (99.7% de los datos)
plt.fill_between(x, y, where=((x >= mu - 3*sigma) & (x <= mu + 3*sigma)), color='salmon', alpha=0.5, label='99.7% (3σ)')

# Configuración del gráfico
plt.title('Ejemplo de la Regla del Intervalo (Regla 68-95-99.7)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
