# Intervalo de confianza
## Regla del intervalo

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parmetros de la distribucin normal
mu = 0  # media
sigma = 1  # desviacin estndar

# Generar datos de una distribucin normal
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = stats.norm.pdf(x, mu, sigma)

# Crear grfico
plt.figure(figsize=(12, 6))

# Graficar la distribucin normal
plt.plot(x, y, label='Distribucin Normal')

# Sombrear el rea de +/- 1 desviacin estndar (68% de los datos)
plt.fill_between(x, y, where=((x >= mu - sigma) & (x <= mu + sigma)), color='skyblue', alpha=0.5, label='68% (1)')

# Sombrear el rea de +/- 2 desviaciones estndar (95% de los datos)
plt.fill_between(x, y, where=((x >= mu - 2*sigma) & (x <= mu + 2*sigma)), color='lightgreen', alpha=0.5, label='95% (2)')

# Sombrear el rea de +/- 3 desviaciones estndar (99.7% de los datos)
plt.fill_between(x, y, where=((x >= mu - 3*sigma) & (x <= mu + 3*sigma)), color='salmon', alpha=0.5, label='99.7% (3)')

# Configuracin del grfico
plt.title('Ejemplo de la Regla del Intervalo (Regla 68-95-99.7)')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```
