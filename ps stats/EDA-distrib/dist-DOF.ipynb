## Degree of freedom. Grados de libertad.

Este código generará una gráfica que ilustra el concepto de grados de libertad en un conjunto de datos, destacando cómo se calculan y su relación con el tamaño de la muestra. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
n = 10  # Número de datos
media_real = 50
datos = np.random.normal(media_real, 5, n)

# Calcular la media muestral
media_muestral = np.mean(datos)

# Grados de libertad
grados_de_libertad = n - 1

# Gráfico
plt.figure(figsize=(10, 6))

# Gráfico de los datos
plt.plot(range(1, n + 1), datos, 'bo-', label='Datos')

# Línea de la media real
plt.axhline(y=media_real, color='r', linestyle='--', label='Media Real')

# Línea de la media muestral
plt.axhline(y=media_muestral, color='g', linestyle='-.', label='Media Muestral')

# Anotación para los grados de libertad
plt.text(n, media_muestral + 2, f'Grados de Libertad (n-1) = {grados_de_libertad}', color='black', fontsize=12, ha='right')

# Títulos y etiquetas
plt.title('Grados de Libertad en un Conjunto de Datos')
plt.xlabel('Índice de Datos')
plt.ylabel('Valor de los Datos')
plt.legend()

# Mostrar gráfico
plt.grid(True)
plt.show()

```
