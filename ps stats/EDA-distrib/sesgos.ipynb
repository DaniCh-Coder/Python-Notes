1. Algunos tipos de sesgo

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo para diferentes tipos de sesgo
np.random.seed(0)

# Sesgo de Selecci√≥n
poblacion_total = 1000
muestra_sesgada = np.random.choice([0, 1], size=poblacion_total, p=[0.8, 0.2])

# Sesgo de Informaci√≥n
datos_reales = np.random.normal(50, 10, 100)
datos_inexactos = datos_reales + np.random.normal(0, 5, 100)

# Sesgo de Confusi√≥n
temperatura = np.random.uniform(20, 40, 100)
consumo_helado = 2 * temperatura + np.random.normal(0, 5, 100)
tasa_ahogamientos = 0.5 * temperatura + np.random.normal(0, 2, 100)

# Graficos
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Grafico de Sesgo de Selecci√≥n
axs[0].bar(['Grupo 0', 'Grupo 1'], [np.sum(muestra_sesgada == 0), np.sum(muestra_sesgada == 1)], color='skyblue')
axs[0].set_title('Sesgo de Selecci√≥n')
axs[0].set_ylabel('Frecuencia')

# Grafico de Sesgo de Informaci√≥n
axs[1].scatter(range(100), datos_reales, color='green', label='Datos Reales')
axs[1].scatter(range(100), datos_inexactos, color='red', alpha=0.5, label='Datos Inexactos')
axs[1].set_title('Sesgo de Informaci√≥n')
axs[1].legend()

# Grafico de Sesgo de Confusi√≥n
axs[2].scatter(consumo_helado, tasa_ahogamientos, color='blue', label='Relaci√≥n Aparente')
axs[2].scatter(temperatura, tasa_ahogamientos, color='red', alpha=0.5, label='Confusi√≥n por Temperatura')
axs[2].set_title('Sesgo de Confusi√≥n')
axs[2].set_xlabel('Consumo de Helado (litros)')
axs[2].set_ylabel('Tasa de Ahogamientos')
axs[2].legend()

# Ajustar dise√±o
plt.tight_layout()
plt.show()
```

2. Distribuciones sesgadas a la izquierda y a la derecha

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generar datos
np.random.seed(0)
n = 1000

# Distribuci√≥n normal centrada
centrada = np.random.normal(loc=0, scale=1, size=n)

# Distribuci√≥n sesgada a la izquierda
sesgada_izquierda = stats.skewnorm.rvs(a=6, loc=0, scale=1, size=n)

# Distribuci√≥n sesgada a la derecha
sesgada_derecha = stats.skewnorm.rvs(a=-6, loc=0, scale=1, size=n)

# Crear gr√°ficos
fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)


# Distribuci√≥n sesgada a la izquierda
axs[0].hist(sesgada_izquierda, bins=30, density=True, alpha=0.6, color='salmon')
xmin, xmax = axs[0].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.skewnorm.pdf(x, 6, loc=0, scale=1)
axs[0].plot(x, p, 'k', linewidth=2)
axs[0].set_title('Distribuci√≥n Sesgada a la Izquierda')

# Distribuci√≥n centrada
axs[1].hist(centrada, bins=30, density=True, alpha=0.6, color='skyblue')
xmin, xmax = axs[1].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, 0, 1)
axs[1].plot(x, p, 'k', linewidth=2)
axs[1].set_title('Distribuci√≥n Normal Centrada')

# Distribuci√≥n sesgada a la derecha
axs[2].hist(sesgada_derecha, bins=30, density=True, alpha=0.6, color='lightgreen')
xmin, xmax = axs[2].get_xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.skewnorm.pdf(x, -6, loc=0, scale=1)
axs[2].plot(x, p, 'k', linewidth=2)
axs[2].set_title('Distribuci√≥n Sesgada a la Derecha')

# Ajustar dise√±o
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos
np.random.seed(0)
n = 1000
data_centrada = np.random.normal(loc=0, scale=1, size=n)
data_sesgada_izquierda = np.random.normal(loc=-1, scale=1, size=n)
data_sesgada_derecha = np.random.normal(loc=1, scale=1, size=n)

# Crear gr√°ficos
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

sns.histplot(data_centrada, kde=True, ax=axs[1], color='skyblue')
axs[1].set_title('Distribuci√≥n Normal Centrada')

sns.histplot(data_sesgada_izquierda, kde=True, ax=axs[0], color='salmon')
axs[0].set_title('Distribuci√≥n Sesgada a la Izquierda')

sns.histplot(data_sesgada_derecha, kde=True, ax=axs[2], color='lightgreen')
axs[2].set_title('Distribuci√≥n Sesgada a la Derecha')

plt.tight_layout()
plt.show()
```
