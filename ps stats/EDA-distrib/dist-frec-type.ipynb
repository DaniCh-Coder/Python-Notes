```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generar la muestra
np.random.seed(0)
muestra = np.random.choice([0, 1, 2, 3, 4, 5, 6], size=30, p=[0.067, 0.167, 0.267, 0.2, 0.167, 0.1, 0.032])

# Crear DataFrame
df = pd.DataFrame(muestra, columns=["Libros Leídos"])

# Calcular frecuencias
frecuencia_absoluta = df["Libros Leídos"].value_counts().sort_index()
frecuencia_relativa = frecuencia_absoluta / frecuencia_absoluta.sum()
frecuencia_acumulada = frecuencia_absoluta.cumsum()
frecuencia_relativa_acumulada = frecuencia_relativa.cumsum()

# Crear un DataFrame de frecuencias
frecuencias = pd.DataFrame({
    "Frecuencia Absoluta": frecuencia_absoluta,
    "Frecuencia Relativa": frecuencia_relativa,
    "Frecuencia Acumulada": frecuencia_acumulada,
    "Frecuencia Relativa Acumulada": frecuencia_relativa_acumulada
})

print(frecuencias)

# Graficar frecuencias
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Frecuencia Absoluta
frecuencias["Frecuencia Absoluta"].plot(kind='bar', ax=axes[0, 0], color='skyblue', title="Frecuencia Absoluta")

# Frecuencia Relativa
frecuencias["Frecuencia Relativa"].plot(kind='bar', ax=axes[0, 1], color='lightgreen', title="Frecuencia Relativa")

# Frecuencia Acumulada
frecuencias["Frecuencia Acumulada"].plot(kind='bar', ax=axes[1, 0], color='salmon', title="Frecuencia Acumulada")

# Frecuencia Relativa Acumulada
frecuencias["Frecuencia Relativa Acumulada"].plot(kind='bar', ax=axes[1, 1], color='plum', title="Frecuencia Relativa Acumulada")

# Ajustar diseño
plt.tight_layout()
plt.show()

```
