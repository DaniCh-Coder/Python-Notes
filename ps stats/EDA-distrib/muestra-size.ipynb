# Muestra y tama√±o
# Ejemplo representativo
###### https://www.chreinvent.com/recursos/cuestiones-de-muestras

```python
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros de la poblaci√≥n
media_poblacion = 50
desviacion_estandar = 10
tamano_poblacion = 10000

# Generar la poblaci√≥n
poblacion = np.random.normal(media_poblacion, desviacion_estandar, tamano_poblacion)

# Tama√±os de muestra
tamanos_muestra = [10, 30, 100, 500]

# Crear subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

# Graficar cada muestra
for i, tamano in enumerate(tamanos_muestra):
    muestra = np.random.choice(poblacion, tamano, replace=False)
    media_muestral = np.mean(muestra)
    
    axs[i].hist(muestra, bins=20, color='skyblue', edgecolor='black')
    axs[i].axvline(media_muestral, color='red', linestyle='dashed', linewidth=2, label=f'Media Muestral = {media_muestral:.2f}')
    axs[i].axvline(media_poblacion, color='green', linestyle='solid', linewidth=2, label=f'Media Poblacional = {media_poblacion:.2f}')
    axs[i].set_title(f'Tama√±o de muestra = {tamano}')
    axs[i].legend()

# Ajustar dise√±o
plt.tight_layout()
plt.show()
```
