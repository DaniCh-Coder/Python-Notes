```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generar datos aleatorios para el ejemplo
np.random.seed(42)
data = np.random.normal(size=100)
data
```

```python

# Crear el gráfico de caja
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, color='skyblue')

# Añadir etiquetas y título
plt.title('Box Plot: Visualización de Cuartiles', fontsize=16)
plt.xlabel('Datos', fontsize=14)
plt.ylabel('Valores', fontsize=14)

# Añadir anotaciones para los cuartiles
plt.text(0, np.percentile(data, 25), 'Q1 (25%)', horizontalalignment='right', fontsize=12, color='blue')
plt.text(0, np.percentile(data, 50), 'Q2 (Mediana, 50%)', horizontalalignment='right', fontsize=12, color='green')
plt.text(0, np.percentile(data, 75), 'Q3 (75%)', horizontalalignment='right', fontsize=12, color='blue')

# Mostrar el gráfico
plt.show()
```

```python
# Datos proporcionados
edades_actrices = [22, 37, 28, 63, 32, 26, 31, 27, 27, 28, 30, 26, 29, 24, 38, 25, 29, 41, 30, 35,
                   35, 33, 29, 38, 54, 24, 25, 46, 41, 28, 40, 39, 29, 27, 31, 38, 29, 25, 35, 60,
                   43, 35, 34, 34, 27, 37, 42, 41, 36, 32, 41, 33, 31, 74, 33, 50, 38, 61, 21, 41,
                   26, 80, 42, 29, 33, 35, 45, 49, 39, 34, 26, 25, 33, 35, 35, 28]

edades_actores = [44, 41, 62, 52, 41, 34, 34, 52, 41, 37, 38, 34, 32, 40, 43, 56, 41, 39, 49, 57,
                  41, 38, 42, 52, 51, 35, 30, 39, 41, 44, 49, 35, 47, 31, 47, 37, 57, 42, 45, 42,
                  44, 62, 43, 42, 48, 49, 56, 38, 60, 30, 40, 42, 36, 76, 39, 53, 45, 36, 62, 43,
                  51, 32, 42, 54, 52, 37, 38, 32, 45, 60, 46, 40, 36, 47, 29, 43]

# Crear el DataFrame
datos = {
    'Edad': edades_actrices + edades_actores,
    'Género': ['Actriz'] * len(edades_actrices) + ['Actor'] * len(edades_actores)
}
df = pd.DataFrame(datos)

# Crear bins para las edades con cut
# https://pandas.pydata.org/docs/reference/api/pandas.cut.html
bins = [20, 30, 40, 50, 60, 70, 80, 90]
labels = ['21–30', '31–40', '41–50', '51–60', '61–70', '71–80', '81–90']
df['Rango de Edad'] = pd.cut(df['Edad'], bins=bins, labels=labels, right=False)
```

## Cálculo de Rangos y Cuartiles
### Cuartiles Q1 (25%), Q2 (50%) (Mediana), Q3 (75%)

```python
# Actrices
dfActriz = df[df['Género'] == 'Actriz']
Q25z=dfActriz['Edad'].quantile(0.25)
Q50z=dfActriz['Edad'].quantile(0.50)
Q75z=dfActriz['Edad'].quantile(0.75)
IQRz = Q75-Q25

# Actores
dfActriz = df[df['Género'] == 'Actor']
Q25r=dfActriz['Edad'].quantile(0.25)
Q50r=dfActriz['Edad'].quantile(0.50)
Q75r=dfActriz['Edad'].quantile(0.75)
IQRr = Q75-Q25
```

```python
print("Actrices                                     Actores")
print(f"Q1 (25th percentile): {Q25z}                   Q1 (25th percentile): {Q25r}")
print(f"Q1 (25th percentile): {Q25z}                   Q1 (25th percentile): {Q25r}")
print(f"Q2 (25th percentile): {Q50z}                   Q2 (25th percentile): {Q50r}")
print(f"Q3 (75th percentile): {Q75z}                  Q3 (75th percentile): {Q75r}")
print(f"Rango intercuartílico (IQR): {IQRz}           Rango intercuartílico (IQR): {IQRr}")
```

Los percentiles/cuartiles de Actores son menores dado que las actrices reciben el Oscar más jóvenes que los hombres.
