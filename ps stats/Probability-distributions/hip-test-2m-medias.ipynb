# Hipótesis relativa a dos medias
###### https://www.statsmodels.org/stable/stats.html#multiple-tests-and-multiple-comparison-procedures
###### https://www.statsmodels.org/stable/generated/statsmodels.stats.weightstats.ttest_ind.html#statsmodels.stats.weightstats.ttest_ind



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind
import statsmodels.stats.weightstats as smw
```

#### Comparar el rendimiento promedio de dos grupos de estudiantes que han utilizado diferentes métodos de estudio:
##### Grupo 1 (método tradicional):
+ Media: 70
+ Desviación estándar: 10
+ Tamaño de la muestra: 50

##### Grupo 2 (método innovador):
+ Media: 75
+ Desviación estándar: 12
+ Tamaño de la muestra: 45

#### Hipótesis:
+ H0: µ1=µ2. No hay diferencia en el rendimiento promedio.
+ H1: µ1≠µ2. Hay una diferencia en el rendimiento promedio.

```python
# Datos del Grupo 1
media1 = 70
std1 = 10
n1 = 50

# Datos del Grupo 2
media2 = 75
std2 = 12
n2 = 45

# Generar muestras simuladas a partir de los parámetros dados
np.random.seed(0)
grupo1 = np.random.normal(media1, std1, n1)
grupo2 = np.random.normal(media2, std2, n2)

# Realizar la prueba t de Student para muestras independientes con scipy
t_stat, p_value = ttest_ind(grupo1, grupo2, equal_var=False)  # Usamos Welch's t-test (varianzas desiguales)

print("Resultados con Scipy:")
print(f'Estadístico t: {t_stat:.4f}')
print(f'Valor p: {p_value:.4f}')

# Realizar la prueba t de Student para muestras independientes con statsmodels
t_stat, p_value, df = smw.ttest_ind(grupo1, grupo2, usevar='unequal')

print("Resultados con Statsmodels:")
print(f'Estadístico t: {t_stat:.4f}')
print(f'Valor p: {p_value:.4f}')
print(f'Grados de libertad: {df:.4f}')

plt.figure(figsize=(10, 6))

plt.hist(grupo1, bins=10, alpha=0.5, label='Grupo 1 (Tradicional)', color='blue')
plt.hist(grupo2, bins=10, alpha=0.5, label='Grupo 2 (Innovador)', color='green')

plt.axvline(np.mean(grupo1), color='blue', linestyle='dashed', linewidth=2)
plt.axvline(np.mean(grupo2), color='green', linestyle='dashed', linewidth=2)

plt.title('Distribución de Rendimiento de Estudiantes')
plt.xlabel('Puntuación')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()

```
