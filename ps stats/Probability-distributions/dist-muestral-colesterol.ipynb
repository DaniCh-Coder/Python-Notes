# Distribución muestral de medias
## Investigación del colesterol en sangre
Supongamos que un investigador quiere estimar la media del colesterol total en la sangre de los adultos de una ciudad. 
+ Debido a que es impráctico medir a todos los adultos, el investigador toma varias muestras de la población y calcula la media de cada muestra.

Estudio

Población y Muestra:
+ Población: Todos los adultos de la ciudad.
+ Tamaño de muestra: n=100n = 100 adultos.

Medición y Cálculo:
+ Se toman 100 muestras de tamaño 100 cada una.
+ Para cada muestra, se calcula la media del colesterol total en la sangre.

Distribución Muestral:
+ Se obtienen 100 medias muestrales.
+ Estas medias muestrales se distribuyen alrededor de la media poblacional del colesterol total.

Según el Teorema del Límite Central, la distribución de estas medias muestrales será aproximadamente normal.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la población
mu = 200  # Media del colesterol total en la población
sigma = 30  # Desviación estándar del colesterol total en la población
tamaño_muestra = 100  # Tamaño de las muestras
n_muestras = 1000  # Número de muestras

# Generar una población normal
np.random.seed(42)
poblacion = np.random.normal(mu, sigma, 10000)

# Calcular las medias muestrales
medias_muestrales = [np.mean(np.random.choice(poblacion, tamaño_muestra)) for _ in range(n_muestras)]

# Gráfico de la distribución muestral de medias
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(medias_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Función de densidad de probabilidad teórica (normal)
media_muestral = np.mean(medias_muestrales)
error_estandar = np.std(medias_muestrales)
x = np.linspace(media_muestral - 4*error_estandar, media_muestral + 4*error_estandar, 100)
pdf = (1 / (error_estandar * np.sqrt(2 * np.pi))) * np.exp(- (x - media_muestral)**2 / (2 * error_estandar**2))
ax.plot(x, pdf, 'r-', lw=2, label='PDF Teórica')

# Etiquetas y título
ax.set_xlabel('Media Muestral del Colesterol')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Muestral de Medias del Colesterol Total')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Histograma de las Medias Muestrales:
+ El histograma muestra la frecuencia de las medias muestrales calculadas a partir de 1000 muestras de tamaño 100 extraídas de la población.
+ Podemos observar una distribución aproximadamente normal centrada en la media poblacional (200200).

Función de Densidad de Probabilidad Teórica (PDF):
+ La línea roja representa la PDF teórica de la distribución normal.
+ La superposición de la PDF teórica sobre el histograma muestra que la distribución muestral de medias sigue la forma esperada según el Teorema del Límite Central.
