```python
import numpy as np
from scipy.stats import hypergeom
import matplotlib.pyplot as plt
```

Los modelos de distribución hipergeométrica extraen objetos de un contenedor. 
+ M es el número total de objetos, (Población)
+ n es el número total de objetos de tipo I. 
+ La variable aleatoria representa el número de objetos de tipo I en N extraídos sin reemplazo de la población total.

## Selección de animales
Supongamos que tenemos una colección de 20 animales, de los cuales 7 son perros. 
+ Entonces, si queremos saber la probabilidad de encontrar una cantidad dada de perros si elegimos al azar 12 de los 20 animales, podemos inicializar una distribución hipergeometrica y representar gráficamente la función de masa de probabilidad:
    + M = 20 animales
    + n = 7 perros
    + x = 12?

##### Función de masa de probabilidad Hipergeométrica

```python
# Construcción de la función de masa de probabilidad con los parámetros dados

# 1. Definición de parámetros
[M, n, N] = [20, 7, 12]     # Se colocan los parametros en una lista antes crear el objeto
x = np.arange(0, n+1)       # Creación de una lista de 1 a la cantidad de perros

# 2. Creación y configuración de la función de masa hipergeométrica
rv = hypergeom(M, n, N)     # Creación de una instancia del objeto con los parametros
pmf_dogs = rv.pmf(x)        # Creación de una lista con los valores de probabilidad

# 3. Impresión de resultados
# 3.a Instancia del objeto
print(f"Función de masa de probabilidad hipergeométrica: hypergeom(M, n, N) ")
print(f"Parametros    : Población, Valores de x, Exitos de la población: {[M, n, N]}")
print(f"Objeto        : {rv}")
# 3.b Valores de x desde 0 hasta el valor buscado
print(f"Valores de x  : {x}")
# lista_redondeada = list(map(lambda x: round(x, 2), lista_original))
# porque pmf_dogs tiene numeros con muchos decimales
n_dogs = list(map(lambda x: round(x, 2), pmf_dogs))
print(f"Probabilidades: {n_dogs}")

# Estadísticos de la función de distribución
print(f"Estadisticos de la Distribución de Probabilidad:")
print(f"Media: {rv.mean()}, Dev.Std.: {rv.std():.2f}, Mediana: {rv.median()}")
print(f"Intervalo de confianza del 95% alrededor de la mediana: {rv.interval(0.95)}")

# 4. Gráfico
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_dogs, 'bo')
ax.vlines(x, 0, pmf_dogs, lw=2)
ax.set_xlabel('# de perros en el grupo de animales tomados como muestra')
ax.set_ylabel('FMP Hipergeométrica')
plt.show()
```

##### Función de masa de probabilidad Hipergeométrica Acumulada

```python
# Construcción de la función de probabilidad acumulada con los parámetros dados

# Creación y configuración de la función de probabilidad acumulada hipergeométrica
# rv = hypergeom(M, n, N)     # Ya esta la, creación de una instancia del objeto con los parametros
cdf_dogs = rv.cdf(x)          # Creación de una lista con los valores de probabilidad de resultados

# 3. Impresión de resultados
# 3.a Instancia del objeto
print(f"Función de masa de probabilidad hipergeométrica acumulada : hypergeom(M, n, N) cdf")
print(f"Parametros    : Población, Valores de x, Exitos de la población: {[M, n, N]}")
print(f"Objeto        : {rv}")
# 3.b Valores de x desde 0 hasta el valor buscado
print(f"Valores de x  : {x}")
# lista_redondeada = list(map(lambda x: round(x, 2), lista_original))
# porque pmf_dogs tiene numeros con muchos decimales
na_dogs = list(map(lambda x: round(x, 2), cdf_dogs))
print(f"Probabilidades Acumuladas: {na_dogs}")

# Gráfico
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, cdf_dogs, 'bo')
ax.vlines(x, 0, cdf_dogs, lw=2)
ax.set_xlabel('# de perros en el grupo de animales tomados como muestra')
ax.set_ylabel('FPA Hipergeométrica')
plt.show()
```
