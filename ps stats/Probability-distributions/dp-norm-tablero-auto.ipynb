```python
# Configuración
import math
import numpy as np
from scipy.stats import norm
from dist_prob_plotgraphs import distr_prob
```

**Contexto**
+ Al diseñar la ubicación de una pantalla de computadora de a bordo en un nuevo modelo de automóviles, 
    + Los ingenieros deben considerar el alcance frontal del conductor. 
    + Si el display se coloca más allá del alcance, el conductor tendría que mover su cuerpo, lo cual lo distraería y resultaría peligroso 
+ Los diseñadores deciden que el reproductor debe ubicarse de manera que esté dentro del alcance del 95% de las mujeres. 
    + Las mujeres tienen alcances frontales distribuidos normalmente, con una media de 27.0 pulgadas y una desviación estándar de 1.3 pulgadas 
        + (según datos de una encuesta antropométrica de Gordon, Churchill, et al.). 

**Analisis**
+ Veamos los estadísticos y la funcón de distribución

```python
# Estadísticos de la distribución uniforme con los datos aportados

# Establecemos los valores de media y desvío estandard de la distribución.
loc=27
scale=1.3

# Vemos los estadísticos de la función de distirbución
mean, var, skew, kurt = norm.stats(loc=loc, scale=scale, moments='mvsk')
sigm = math.sqrt(var)
print(f"media: {mean}, desv.std.: {sigm} , sesgo: {skew}, kurtosis: {kurt})")
distr_prob(norm(loc=loc, scale=scale))
```

**Interpretación**
+ Las personas bajo estudio tienen un alcance medio de 27 pulgadas con una desviación estandar de 1.3 pulgadas.
+ No es probable que se encuentren personas con alcance menor a 22 pulgadas o mayor a 32 pulgadas.

a. Calcule el alcance frontal de las mujeres que separa al 95% superior del resto.

```python
# a. Calcule el alcance frontal de las mujeres que separa al 95% superior del resto.

print(f"Estadísticos de la distribución: media: {mean}, desv.std.: {sigm} , sesgo: {skew}, kurtosis: {kurt})")

# Establecemos el percenitl que estamos buscando
percentil = 1- 0.95 # Porque estamos buscando el 95% superior en este caso.

# Establecemos el valor de x o z (alcance) correspondiente al percentil buscado
alcance = norm(loc=mean, scale=sigm).ppf(percentil)    # Este es el valor buscado
print(f"El alcance separa al 99.5% de las mujeres con mayor alcance del 0.5% de las de menor alcance es: {alcance}")
p95 = norm(loc=mean, scale=sigm).pdf(alcance)
print(f"El valor exacto de probabilidad para este alcance ({alcance}) es: {p95}")

# Comprobación-Demostración
# Calculamos la probabilidad acumulada correspondiente a ese valor de x (alcance)
p= norm.cdf(alcance, loc=mean, scale=sigm)

print(f"Probabilidad de encontrar una mujer con un alcance menor a: {alcance} es: {p}")
distr_prob(norm(loc=mean, scale=sigm), b=alcance)

print(f"Probabilidad de encontrar un mujer con un alcance mayor a: {alcance} es: {1-p}")
distr_prob(norm(loc=mean, scale=sigm), b=alcance, between=False)
```

**Interpretación**
+ El alcance frontal de 24.9 pulgadas (redondeado) separa al 95% superior del resto, 
    + ya que el 95% de las mujeres tienen alcances frontales mayores que 24.9 pulgadas, 
    + y el 5% de las mujeres tienen alcances frontales menores que 24.9 pulgadas.
