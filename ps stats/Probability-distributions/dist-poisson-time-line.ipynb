### Ejemplo: Llegada de Clientes a un Centro de Llamadas
Supongamos que estamos modelando la llegada de llamadas a un centro de atención al cliente. 
+ Sabemos que, en promedio, llegan 5 llamadas por minuto. 
+ Queremos utilizar la distribución de Poisson para modelar el número de llamadas que llegan en un minuto y analizar los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parámetros de la distribución de Poisson
lambda_ = 5  # Tasa promedio de llegada de llamadas por minuto

# Valores posibles del número de llamadas en un minuto
x = np.arange(0, 20)

# Función de masa de probabilidad (PMF)
pmf = poisson.pmf(x, lambda_)

# Simulación de datos
n_simulaciones = 1000
simulaciones = poisson.rvs(lambda_, size=n_simulaciones)

# Gráfico de la distribución de Poisson
fig, ax = plt.subplots(figsize=(10, 6))

# Histogramas y PMF
ax.hist(simulaciones, bins=np.arange(20) - 0.5, density=True, alpha=0.6, color='g', edgecolor='black', label='Simulaciones')
ax.plot(x, pmf, 'bo', ms=8, label='PMF Teórica')
ax.vlines(x, 0, pmf, colors='b', lw=2)

# Etiquetas y título
ax.set_xlabel('Número de Llamadas por Minuto')
ax.set_ylabel('Probabilidad')
ax.set_title('Distribución de Poisson: Llegada de Llamadas a un Centro de Atención')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
