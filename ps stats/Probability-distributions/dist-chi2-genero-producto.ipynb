# Distribución Chi-2
## Relación entre género y preferencia por productos A o B

Supongamos que queremos evaluar si hay una relación entre el género (masculino o femenino) y la preferencia por un tipo de producto (A o B) en una tienda. 
+ Los datos observados se resumen en la siguiente tabla de contingencia. 
    + Producto A      Producto B     Total
    + Masculino          30                      20               50
    + Femenino           25                      25               50
    + Total              55                      45              100

Queremos usar la prueba chi-cuadrado para determinar si las preferencias de productos son independientes del género. Este ejemplo se calcula en python con un jupyter notebook compartido en github.

```python
import scipy.stats as stats
import numpy as np

# Datos observados
observed = np.array([[30, 20],
                     [25, 25]])

# Realizar la prueba chi-cuadrado
chi2, p, dof, expected = stats.chi2_contingency(observed)

# Resultados
print(f'Estadístico Chi-cuadrado: {chi2:.2f}')
print(f'P-valor: {p:.4f}')
print(f'Grados de Libertad: {dof}')
print('Frecuencias Esperadas:')
print(expected)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros de la distribución chi-cuadrado
df = 1  # Grados de libertad
alpha = 0.05  # Nivel de significancia
chi2_observado = 0.65  # Estadístico chi-cuadrado observado

# Valores críticos de chi-cuadrado
chi2_critico = stats.chi2.ppf(1 - alpha, df)

# Rango de valores para la gráfica
x = np.linspace(0, 10, 1000)
y = stats.chi2.pdf(x, df)

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar la distribución chi-cuadrado
ax.plot(x, y, 'b-', label=f'Chi-cuadrado (df={df})')

# Sombrear la zona crítica
x_critico = np.linspace(chi2_critico, 10, 1000)
y_critico = stats.chi2.pdf(x_critico, df)
ax.fill_between(x_critico, y_critico, alpha=0.5, color='red', label='Zona crítica')

# Añadir líneas verticales para el valor crítico y el estadístico observado
ax.axvline(chi2_critico, color='red', linestyle='--', label=f'Valor crítico = {chi2_critico:.2f}')
ax.axvline(chi2_observado, color='green', linestyle='--', label=f'Estadístico observado = {chi2_observado:.2f}')

# Etiquetas y título
ax.set_xlabel('Valores de Chi-cuadrado')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title(f'Distribución Chi-cuadrado con {df} Grado de Libertad')
ax.legend()

# Mostrar el gráfico
plt.show()
```
