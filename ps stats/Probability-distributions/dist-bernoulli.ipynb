# Distribución de Bernoulli

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución de Bernoulli
p = 0.3  # Probabilidad de éxito
q = 1 - p  # Probabilidad de fracaso
x = np.array([0, 1])  # Resultados posibles: 0 (fracaso), 1 (éxito)
probabilidades = [q, p]  # Probabilidades correspondientes

# Configuración de la figura y el subplot
fig, ax = plt.subplots(figsize=(8, 6))
bar_colors = ['#ff9999','#66b3ff']  # Colores para las barras

# Graficar las barras de la distribución de Bernoulli
bars = ax.bar(x, probabilidades, color=bar_colors, alpha=0.6, width=0.4)

# Añadir etiquetas y título
ax.set_xticks(x)
ax.set_xticklabels(['Fracaso (0)', 'Éxito (1)'])
ax.set_xlabel('Resultado')
ax.set_ylabel('Probabilidad')
ax.set_title('Distribución de Bernoulli (p = 0.3)')

# Añadir anotaciones a las barras
for bar, prob in zip(bars, probabilidades):
    height = bar.get_height()
    ax.annotate(f'{prob:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 puntos de offset vertical
                textcoords="offset points",
                ha='center', va='bottom')

# Añadir elementos visuales creativos
# Añadimos flechas para enfatizar las probabilidades
for i, (label, prob) in enumerate(zip(['Fracaso', 'Éxito'], probabilidades)):
    ax.annotate(f'{label}\n{prob*100:.0f}%',
                xy=(i, prob),
                xytext=(i, prob + 0.1),
                arrowprops=dict(facecolor='black', shrink=0.05),
                ha='center', va='center')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
