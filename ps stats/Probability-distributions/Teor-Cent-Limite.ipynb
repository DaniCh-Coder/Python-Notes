# Teorema Central del Límite
### Comprobación
+ Construimos una población grande de numeros naturales enteros del 0 al 9.
    + Observamos que tiene una distribución uniforme.
+ Tomamos una cantidad grande de muestras sobre esa población.
    + Construimos la distribución de las medias de esa muestra.
+ Observamos que la distribución de las medias tiene una forma normal.
+ Coomprobamos que la media de la distribución de las medias tiene:
    + Una media igual a la media de la población.
    + Una desvíación estandard igual al desvío estandar de la media sobre la raiz del tamaño de la muestra.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Funciones de gráficos
# Configuración del histograma
def plot_poblacion(poblacion, t_p):
    plt.figure(figsize=(8, 5))
    hist, bins, patches = plt.hist(poblacion, bins=np.arange(11) - 0.5, edgecolor='black', rwidth=0.8)

    # Configurar los ejes y el título
    plt.title(f'Histograma de {t_p} de números aleatorios entre 0 y 9', fontsize=15)
    plt.xlabel('Número', fontsize=12)
    plt.ylabel('Frecuencia', fontsize=12)
    plt.xticks(range(10))  # Alineación correcta de los números naturales
    plt.grid(axis='y', alpha=0.75)

    # Mostrar el histograma
    plt.tight_layout()
    plt.show()

def plot_med_muest(medias_muestrales, c_m, t_m):   
    plt.figure(figsize=(8, 6))
    plt.hist(medias_muestrales, bins=50, edgecolor='black', color='skyblue')
    plt.title(f'Histograma de las Medias Muestrales. Cantidad Muestras: {c_m}, Tamaño Muestras: {t_m}', fontsize=12)
    plt.xlabel('Media Muestral', fontsize=12)
    plt.ylabel('Frecuencia', fontsize=12)
    plt.grid(axis='y', alpha=0.75)
    plt.tight_layout()
    plt.show()

# Función para tomar muestras y calcular sus medias
def calcular_medias_muestras(poblacion, num_muestras, tamano_muestra):
    # Crear un array para almacenar las muestras
    print(f"poblacion: {poblacion}, num_muestras: {num_muestras}, tamano_muestra: {tamano_muestra}")
    muestras = np.zeros((num_muestras, tamano_muestra), dtype=float)
        # Rellenar el array de muestras con muestras aleatorias de la población
    for i in range(num_muestras):
        muestras[i] = np.random.choice(poblacion, tamano_muestra, replace=False)
    medias_muestrales = np.mean(muestras, axis=1)
    return medias_muestrales

def print_pobl_y_muestra(poblacion, ):
    print("Población generada:")
    print(poblacion[:100])  # Imprimir las primeras 100 entradas para verificar  
    print("\nPrimera muestra:")
    print(muestras[0])  # Imprimir la primera muestra para verificar
    print("\nForma del array de muestras:", muestras.shape)
    
def tcl(poblacion, t_p, t_m):
    # Calcula cantidad de meustras
    c_m = t_p // t_m

    # Paso 2 y 3: Construcción de las muestras y cálculo de media de las muestras.
    medias_muestrales = calcular_medias_muestras(poblacion, c_m, t_m)

    # Paso 4: Dibujar el histograma de las medias muestrales
    plot_med_muest(medias_muestrales, c_m, t_m)
    
```

```python
t_p = 50000
# Generar una población con una distribución no normal (distribución uniforme en este caso)
poblacion = np.random.randint(0, 10, t_p)

plot_poblacion(poblacion, t_p)

ts_ms = [t_p, t_p//5, t_p//10, t_p//50, t_p//100, t_p//500, t_p//1000, t_p//5000,  t_p//10000, t_p//t_p]
ts_ms = ts_ms[::-1]

for t_m in ts_ms:
    tcl(poblacion, t_p,t_m)
```
