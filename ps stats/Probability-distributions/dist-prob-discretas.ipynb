# Distribuciones de probabilidad discretas
## Ejemplos gr√°ficos

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Par√°metros para las distribuciones
n = 10  # N√∫mero de ensayos para la Binomial y la Binomial Negativa
p = 0.5  # Probabilidad de √©xito
k = 20  # N√∫mero de ensayos para la Hipergeom√©trica
M = 50  # N√∫mero total de objetos
N = 10  # N√∫mero de objetos de inter√©s
low, high = 1, 10  # L√≠mites para la Uniforme Discreta
lambd = 3  # Tasa para la Poisson
prob_multinomial = [0.2, 0.3, 0.5]  # Probabilidades para la Multinomial

# Configuraci√≥n de la figura y los subplots
fig, axs = plt.subplots(2, 4, figsize=(20, 10))
fig.suptitle('Distribuciones Discretas', fontsize=20)

# Distribuci√≥n Binomial
x = np.arange(0, n+1)
axs[0, 0].bar(x, stats.binom.pmf(x, n, p), color='b', alpha=0.6)
axs[0, 0].set_title('Binomial')

# Distribuci√≥n Multinomial (ejemplo con 3 categor√≠as y 10 ensayos)
multinomial_data = stats.multinomial.rvs(n=n, p=prob_multinomial, size=1000)
multinomial_counts = np.apply_along_axis(np.bincount, 1, multinomial_data, minlength=n+1)
multinomial_means = multinomial_counts.mean(axis=0)
axs[0, 1].bar(np.arange(n+1), multinomial_means, color='g', alpha=0.6)
axs[0, 1].set_title('Multinomial')

# Distribuci√≥n Geom√©trica
x = np.arange(1, n+1)
axs[0, 2].bar(x, stats.geom.pmf(x, p), color='r', alpha=0.6)
axs[0, 2].set_title('Geom√©trica')

# Distribuci√≥n Binomial Negativa
x = np.arange(0, 2*n)
axs[0, 3].bar(x, stats.nbinom.pmf(x, n, p), color='c', alpha=0.6)
axs[0, 3].set_title('Binomial Negativa')

# Distribuci√≥n Hipergeom√©trica
x = np.arange(0, n+1)
axs[1, 0].bar(x, stats.hypergeom.pmf(x, M, N, k), color='m', alpha=0.6)
axs[1, 0].set_title('Hipergeom√©trica')

# Distribuci√≥n Uniforme Discreta
x = np.arange(low, high+1)
axs[1, 1].bar(x, stats.randint.pmf(x, low, high+1), color='y', alpha=0.6)
axs[1, 1].set_title('Uniforme Discreta')

# Distribuci√≥n de Poisson
x = np.arange(0, 2*lambd+1)
axs[1, 2].bar(x, stats.poisson.pmf(x, lambd), color='k', alpha=0.6)
axs[1, 2].set_title('Poisson')

# Distribuci√≥n de Bernoulli
x = np.array([0, 1])
axs[1, 3].bar(x, stats.bernoulli.pmf(x, p), color='orange', alpha=0.6)
axs[1, 3].set_title('Bernoulli')

# Ajustes finales
for ax in axs.flat:
    ax.set_xlabel('Valores')
    ax.set_ylabel('Probabilidad')
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
