## Ejemplo de paquetes de envío
Supongamos que queremos estudiar el peso promedio de los paquetes enviados por una empresa de mensajería. 
+ La distribución del peso de todos los paquetes puede tener una forma desconocida o ser sesgada. 
+ Sin embargo, si tomamos muestras suficientemente grandes de estos paquetes, la distribución de las medias muestrales de peso será aproximadamente normal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una población con una distribución no normal (distribución exponencial)
np.random.seed(42)
poblacion = np.random.exponential(scale=2, size=10000)

# Calcular las medias muestrales
tamaño_muestra = 50  # Tamaño de las muestras
n_muestras = 1000  # Número de muestras
medias_muestrales = [np.mean(np.random.choice(poblacion, tamaño_muestra)) for _ in range(n_muestras)]

# Gráfico de la distribución muestral de medias
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(medias_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Función de densidad de probabilidad teórica (normal)
media_muestral = np.mean(medias_muestrales)
error_estandar = np.std(medias_muestrales)
x = np.linspace(media_muestral - 4*error_estandar, media_muestral + 4*error_estandar, 100)
pdf = (1 / (error_estandar * np.sqrt(2 * np.pi))) * np.exp(- (x - media_muestral)**2 / (2 * error_estandar**2))
ax.plot(x, pdf, 'r-', lw=2, label='PDF Teórica')

# Etiquetas y título
ax.set_xlabel('Media Muestral')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Demostración del Teorema del Límite Central')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
