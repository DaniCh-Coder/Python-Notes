### Ejemplo: componente defectuoso
Supongamos que la probabilidad de un componente de computadora defectuoso es de 0.2.
+ Calcule la probabilidad de que el primer defecto se descubra en el séptimo componente probado.
     + P(X = k) = P(x)= p(1-p)^((k-1))

Al analizar la formula y el ejemplo se puede ver que, como no se tiene en cuenta el tamaño de la muestra, es decir, 
+ la muestra es siempre uno y no se considera aquí la ley de los grandes números. En efecto, al ver la formula vemos que:
    1. cuanto mayor es la probabildad de éxtio, mayor es la probabilidad frente a menos intentos y viceversa.
    2. para un solo intento, la probabilidad solo por el producto de ambas probabilidades.

Este modelo nos dice que frente a escases de datos (aquí no tenemos tamaño de población ni de muestra) cada vez es más dificil descubrir algo.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
p = 0.2  # Probabilidad de defecto
k = 7    # Número del intento

# Cálculo de la probabilidad
probability = (1 - p)**(k - 1) * p

# Mostrar el resultado
print(f"La probabilidad de que el primer defecto se descubra en el séptimo componente es {probability:.4f}")

# Generar datos para un gráfico representativo
x = np.arange(1, 15)  # Números de intentos (k)
y = (1 - p)**(x - 1) * p  # Probabilidades de cada intento

# Crear el gráfico
plt.figure(figsize=(8, 5))
plt.bar(x, y, color='lightblue', edgecolor='black', alpha=0.7, label="Distribución geométrica")
plt.axvline(k, color='red', linestyle='--', label=f"Primer defecto en k={k} (P={probability:.4f})")

# Etiquetas y leyenda
plt.title("Distribución geométrica: Probabilidad del primer defecto", fontsize=14)
plt.xlabel("Número de componentes probados (k)")
plt.ylabel("Probabilidad")
plt.xticks(x)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

### Tiempo hasta la Primera Venta Exitosa
Imagina que trabajas en un equipo de ventas que hace llamadas telefónicas a clientes potenciales. 
+ La probabilidad de realizar una venta exitosa en cada llamada es del 10%. 
+ Queremos saber cuántas llamadas es probable que haga un vendedor antes de conseguir su primera venta exitosa.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Parámetros de la distribución geométrica
p = 0.1  # Probabilidad de éxito en cada llamada
n = 1000  # Número de simulaciones

# Simulación del número de llamadas hasta la primera venta exitosa
simulaciones = np.random.geometric(p, n)

# Gráfico de la distribución geométrica
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(1, max(simulaciones) + 1)
pmf = geom.pmf(x, p)

# Histogramas y PMF
ax.hist(simulaciones, bins=x, density=True, alpha=0.6, color='g', edgecolor='black', label='Simulaciones')
ax.plot(x, pmf, 'bo', ms=8, label='PMF Teórica')
ax.vlines(x, 0, pmf, colors='b', lw=2)

# Etiquetas y título
ax.set_xlabel('Número de llamadas hasta la primera venta exitosa')
ax.set_ylabel('Probabilidad')
ax.set_title('Distribución Geométrica: Tiempo hasta la Primera Venta Exitosa')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
