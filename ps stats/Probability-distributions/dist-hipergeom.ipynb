### Ejemplo: Selección de Pelotas de una Bolsa

Supongamos que tenemos una bolsa con 20 pelotas, de las cuales 7 son rojas (éxitos) y 13 son azules (fracasos). 
+ Queremos saber la probabilidad de extraer 4 pelotas rojas al seleccionar 10 pelotas sin reemplazo.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Parámetros de la distribución hipergeométrica
N = 20  # Tamaño de la población (total de pelotas)
K = 7  # Número de éxitos en la población (pelotas rojas)
n = 10  # Número de extracciones (pelotas extraídas)

# Valores posibles de k (número de pelotas rojas extraídas)
x = np.arange(0, n+1)

# Probabilidad de obtener k pelotas rojas
pmf = hypergeom.pmf(x, N, K, n)

# Simulación de la distribución hipergeométrica
simulaciones = hypergeom.rvs(N, K, n, size=1000)

# Gráfico de la distribución hipergeométrica
fig, ax = plt.subplots(figsize=(10, 6))

# Histogramas y PMF
ax.hist(simulaciones, bins=np.arange(n+2) - 0.5, density=True, alpha=0.6, color='g', edgecolor='black', label='Simulaciones')
ax.plot(x, pmf, 'bo', ms=8, label='PMF Teórica')
ax.vlines(x, 0, pmf, colors='b', lw=2)

# Etiquetas y título
ax.set_xlabel('Número de Pelotas Rojas Extraídas')
ax.set_ylabel('Probabilidad')
ax.set_title('Distribución Hipergeométrica: Extracciones Sin Reemplazo de Pelotas')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
