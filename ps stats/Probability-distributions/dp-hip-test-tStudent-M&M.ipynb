+ **La desviaci칩n est치ndard de todos los 465 dulces es desconocido**
+ A continuaci칩n se presentan los pesos (en gramos), los cuales tienen una media de x = 0.8635 y una desviaci칩n est치ndar de s = 0.0576 g. 
    + 0.751, 먝0.841, 먝0.856, 먝0.799, 먝먝0.966,    0.859,   0.857, 먝0.942, 먝0.873, 먝0.809, 먝0.890, 먝0.878, 먝0.905
+ En el empaque se afirma que el peso neto del contenido es 396.9 g, de manera que los M&M deben tener un peso medio de al menos 396.9/465 = 0.8535 g 
    + para dar la cantidad anunciada. 

+ Utilice los datos muestrales con un nivel de significancia de 0.05, 
    + para probar la aseveraci칩n que hizo un gerente de producci칩n de que los M&M tienen en realidad una media mayor que 0.8535 g, 
    + de manera que los consumidores est치n recibiendo m치s que la cantidad indicada en la etiqueta.

## An치lisis de la muestra

```python
mmsample = [0.751, 0.841, 0.856, 0.799, 0.966, 0.859, 0.857, 0.942, 0.873, 0.809, 0.890, 0.878, 0.905]
```

1. La muestra es aleatoria simple.
2. Se **desconoce** el valor de la desviaci칩n est치ndar poblacional 洧룩.
3. Se satisface una o ambas de las siguientes condiciones:

Para esto es preciso observar a) Estad칤sticos de la muestra y b) Normalidad

## Estad칤sticos de la muestra

```python
me, sd = mean(mmsample), stdev(mmsample)
mi, ma = min(mmsample), max(mmsample)
print(f"media:{me}, desv칤o estandard:{sd}, minimo:{mi}")
```

```python
# Configuraci칩n de los gr치ficos
plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(mmsample, bins=7, edgecolor='black', alpha=0.7)
plt.title('Histograma de mmsample')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(mmsample, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot de mmsample')
plt.xlabel('Valor')

# Mostrar gr치ficos
plt.tight_layout()
plt.show()

print(f"\nEstad칤sticos en orden para ver en grafica: Observaciones: {len(mmsample)}, Rango: {(ma - mi):.3f}.")
# Medidas de tendencia central
print(f"min: {mi},   -1sd: {-sd:.3f},   mode: {mode(mmsample)},   median: {median(mmsample)},"\
      f"   media: {me:.3f},   +1sd: {sd:.3f},   max: {ma}")
import numpy as np

def calculate_iqr_and_boxplot_limits(data):
  """
  This function calculates the interquartile range (IQR), quartiles, 
  and upper and lower limits for a boxplot.

  Args:
      data: A list or NumPy array of numerical data.

  Returns:
      A dictionary containing the following keys:
          - 'Q1': The first quartile (25th percentile).
          - 'Q3': The third quartile (75th percentile).
          - 'IQR': The interquartile range (IQR).
          - 'lower_limit': The lower limit for the boxplot.
          - 'upper_limit': The upper limit for the boxplot.
  """

  # Sort the data in ascending order
  data = np.sort(data)

  # Calculate the number of data points
  n = len(data)

  # Check for empty data
  if n == 0:
    return {
      'Q1': None,
      'Q3': None,
      'IQR': None,
      'lower_limit': None,
      'upper_limit': None,
    }

  # Find the quartiles based on even or odd number of data points
  if n % 2 == 0:
    Q1_index = int(n / 4) - 1
    Q3_index = int(n / 4 * 3) - 1
  else:
    Q1_index = int((n + 1) / 4) - 1
    Q3_index = int((n + 1) / 4 * 3) - 1

  # Calculate the quartiles
  Q1 = data[Q1_index]
  Q3 = data[Q3_index]

  # Calculate the interquartile range
  IQR = Q3 - Q1

  # Calculate the upper and lower limits
  lower_limit = Q1 - 1.5 * IQR
  upper_limit = Q3 + 1.5 * IQR

  return {
    'Q1': Q1,
    'Q3': Q3,
    'IQR': IQR,
    'lower_limit': lower_limit,
    'upper_limit': upper_limit,
  }

# Example usage
data = mmsample
results = calculate_iqr_and_boxplot_limits(data)

print(f"First quartile (Q1): {results['Q1']}")
print(f"Third quartile (Q3): {results['Q3']}")
print(f"Interquartile range (IQR): {results['IQR']:.3f}")
print(f"Lower limit: {results['lower_limit']:.3f}")
print(f"Upper limit: {results['upper_limit']:.3f}")
```

### An치lisis de normalidad de los datos de la muestra

```python
# Pruebas de normalidad
shapiro_test = stats.shapiro(data)
anderson_test = stats.anderson(data, dist='norm')
dagostino_test = stats.normaltest(data)

print("Shapiro-Wilk Test")
print("Eval칰a la hip칩tesis nula de que una muestra proviene de una distribuci칩n normal. Es una prueba muy sensible para muestras peque침as.")
print("Shapiro-Wilk Test:", shapiro_test)
print("Si el valor p es menor que 0.05, se rechaza la hip칩tesis nula, sugiriendo que los datos no son normales.\n")

print("Anderson-Darling Test")
print("Eval칰a la hip칩tesis nula de que los datos siguen una distribuci칩n espec칤fica (por defecto, normal).")
print("Anderson-Darling Test:", anderson_test)
print("El resultado incluye un estad칤stico de prueba y una serie de valores cr칤ticos.\n"
      f"Si el estad칤stico es mayor que los valores cr칤ticos correspondientes a ciertos niveles de significancia, se rechaza la hip칩tesis nula.\n")

print("D'Agostino's K-squared Test")
print("Compara una muestra con una distribuci칩n de referencia (normal en este caso).")
print("D'Agostino's K-squared Test:", dagostino_test)
print("Si el valor p es menor que 0.05, se rechaza la hip칩tesis nula, sugiriendo que los datos no son normales.\n")

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, loc=np.mean(data), scale=np.std(data))
plt.plot(x, p, 'k', linewidth=2)
plt.title("Histograma y curva normal")

plt.subplot(1, 2, 2)
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q Plot")

plt.show()

# Ajuste de distribuci칩n con Fitter
f = Fitter(data, distributions=['norm', 'lognorm', 'expon', 'gamma'])
f.fit()
f.summary()
```

## Distribuci칩n t-Student de la Poblaci칩n
Debe notarse que se analiza la media muestral en relaci칩n a la poblaci칩n.
+ La poblaci칩n en este caso ser칤an los 465 dulces
    + En el empaque se afirma que el peso neto del contenido es 396.9 g, de manera que los M&M deben tener un peso medio de al menos 396.9/465 = 0.8535 g 
    + para dar la cantidad anunciada.
+ Se sabe tambi칠n (por que se ha calculado) que la desviaci칩n est치ndard de todos los 465 dulces de 0.0565 

```python
# Distribuci칩n t-Student
mep, sdp = 396.9/465, 0.0565 
df = 12
fig, ax = distr_prob(stats.t(df=df, loc=mep, scale=sdp))
```

## Hip칩tesis del gerente de producci칩n
+ Probar la aseveraci칩n que hizo un gerente de producci칩n de que los M&M tienen en realidad una media mayor que 0.8535 g, 
    + de manera que los consumidores *est치n recibiendo m치s que la cantidad indicada en la etiqueta*.
## Prueba de hip칩tesis
+ Para probar la hip칩tesis que los consumidores *est치n recibiendo m치s que la cantidad indicada en la etiqueta*
    + comparamos la media de la con la media de la poblaci칩n y
    + vemos d칩nde se ubica la media de la muestra en comparaci칩n con la regi칩n cr칤tica.
## H0 y H1
+ Definimos
    + Hip칩tesis Nula H0: 췃 = 0.854
    + Hip칩tesis Alternativa: 췃 > 0.854

## Analisis del intervalo de confianza de la poblaci칩n
+ El IC nos dice con una probabilidad del 95% el rango de valores posibles para la media de la poblaci칩n
+ Es un primer an치lisis aproximado:
    + Si el valor de la media de la muestra entre dentro del intevalo de confianza entonces es un valor posible por azar

```python
# An치lisis de la Regi칩n Cr칤tica con t-Student
# grados de libertad n-1 y estad칤stico de prueba
df= 12
z = (me - mep) / (sd / math.sqrt(13))
print(f"Tama침o de la muestra: {len(mmsample)}, Grados de libertad: {df}")
print(f"Estad칤stico de la muestra: {z}")

# Gr치fico de la distribuci칩n t-student con df grados de libertad
fig, ax = reg_critica(stats.t(df=df))
                      
# dibuja la l칤nea de la media de la poblaci칩n
ax.plot([0, 0], [0, stats.t.pdf(x=0, df=df)], linestyle="dashed")  # dibuja la l칤nea punteada
ax.text(x=0, y=-0.02, s=f'z={0}', ha='center', va='bottom', fontsize=8, color='g')   # A침ade valor

# dibuja la l칤nea de la media de la muestra
ax.plot([z, z], [0, stats.t.pdf(x=z, df=df)], linestyle="dashed", color='red')  # dibuja la l칤nea punteada
ax.text(x=z, y=-0.02, s=f'z={z:.3f}', ha='left', va='bottom', color='red', fontsize=8)   # A침ade valor estadistico de prueba

plt.show(fig)
```

**Observaciones**
+ La media de la muestra est치 dentro del IC intervalo de confianza del 95%
+ La media de la muestra esta proxima a la media de la poblaci칩n 

### **Conclusiones del an치lisis del intervalo de confianza**
+ Es muy probable que 췃 = 0.864 haya aparecido como consecuencia del azar
+ Observando el intervalo de confianza la H0 parece aceptable

## An치lisis del estad칤stico de prueba t y la regi칩n cr칤tica
+ Es un an치lisis m치s exahustivo que el anterior
    + Si el estad칤stico de la muestra entra dentro de la regi칩n cr칤tica la H0 se rechaza
    + Si el estad칤stico de la muestra est치 fuera de la regi칩n cr칤tica la H0 no se puede rechazar

```python
# An치lisis del estad칤stico 췃
df=12
t = (me - mep) / (sd / math.sqrt(13))
print(f"Estad칤stico t:{z}")
fig, ax = reg_critica(stats.t(df=df), side='right')

# dibuja la l칤nea de la media de la poblaci칩n
ax.plot([0, 0], [0, stats.t.pdf(x=0, df=df)], linestyle="dashed")  # dibuja la l칤nea punteada
ax.text(x=0, y=-0.0, s=f't={0}', ha='center', va='bottom', fontsize=8, color='g')   # A침ade valor de x de la muestra

# dibuja la l칤nea de la media de la muestra
ax.plot([t, t], [0, stats.t.pdf(x=t, df=df)], linestyle="dashed", color='red')  # dibuja la l칤nea punteada
ax.text(x=z, y=-0.02, s=f't={t:.3f}', ha='left', va='bottom', color='red', fontsize=8)   # A침ade valor de x de la muestra

plt.show(fig)
```

**Observaciones**
+ La media de la muestra est치 fuera de la zona cr칤tica del nivel de signifiancia alfa marcada en lila
+ La media de la muestra esta proxima a la media de la poblaci칩n y fuera de la zona cr칤tica

### **Conclusiones del an치lisis del estad칤stico de prueba (Media)**
+ Es muy probable que 췃 = 0.864 haya aparecido como consecuencia del azar

### **An치lisis del valor P**
+ Con el estad칤stico de prueba se puede analizar el valor P en contraste con 풤

```python
# Analisis del valor P con t-Student
df=12
fig, ax = distr_prob(stats.t(df=df, loc=0, scale=1), a=z)
```

**Observaciones**
+ El valor P asociado al estad칤stico de prueba 췃 = 0.864 es mucho mayor que 풤
    + Es la regi칩n sobreada de la figura: 42.7% >> 5%
### **Conclusiones del valor P del estad칤stico de Prueba**
+ El valor P es tan alto que hace aceptable la Hip칩tesis Nula H0
+ No se puede rechazar la H0: 췃 = 0.854

## **Conclusiones finales**
