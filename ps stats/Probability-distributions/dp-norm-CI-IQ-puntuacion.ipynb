```python
import math
import numpy as np
from scipy.stats import norm
from dist_prob_plotgraphs import distr_prob, inter_confi
```

+ Un psicólogo está diseñando un experimento para probar la eficacia de un nuevo programa de capacitación para vigilantes de seguridad aeroportuaria. 
+ Desea comenzar con un grupo homogéneo de sujetos con puntuaciones de CI comprendidas entre 85 y 125. 
+ Dado que las puntuaciones de CI se distribuyen normalmente, con una media de 100 y una desviación estándar de 15.

```python
# Estadísticos de la distribución uniforme con los datos aportados

# Establecemos los valores de media y desvío estandard de la distribución.
loc=100
scale=15

# Vemos los estadísticos de la función de distirbución
mean, var, skew, kurt = norm.stats(loc=loc, scale=scale, moments='mvsk')
sigm = math.sqrt(var)
print(f"media: {mean}, desv.std.: {sigm} , sesgo: {skew}, kurtosis: {kurt})")
distr_prob(norm(loc=loc, scale=scale))
```

**Interpretación** 
+ Las personas tienen un coeficiente intelectual promedio de 100, con una distribución estandar de 15.
      + La región sombreada representa a las personas con CI entre 40 y 160 con su respectiva probabilidad.

#### Rango de CI amplio
a. ¿qué porcentaje de la gente tiene una puntuación de CI entre 85 y 125?

```python
# a. ¿qué porcentaje de la gente tiene una puntuación de CI entre 85 y 125?
min = 85
max = 125

print(f"media: {mean}, desv.std.: {sigm} , sesgo: {skew}, kurtosis: {kurt})")
p1= norm.cdf(max, loc=mean, scale=sigm)
p0= norm.cdf(min, loc=mean, scale=sigm)
p = p1-p0

print(f"Probabilidad acumulada {max}: {p1}")
distr_prob(norm(loc=mean, scale=sigm), b=max)

print(f"Probabilidad acumulada {min}: {p0}")
distr_prob(norm(loc=mean, scale=sigm), b=min)

print(f"Probabilidad de encontrar una persona con un CI entre {min} y {max}: {p}")
distr_prob(norm(loc=mean, scale=sigm), a=min, b=max)

print(f"Probabilidad de NO encontrar una persona con un CI entre {min} y {max}: {1-p}")
distr_prob(norm(loc=mean, scale=sigm), a=min, b=max, between=False)
```

**Interpretación**
+ El 79.4% de la población tiene un coeficiente entre 85 y 125.
+ Se infiere esto porque se observa que:
    + el 95.2% de las personas tiene un CI menor a 125
    + el 15.9% de las personas tiene un CI menor a 85
+ La diferencia entre ambos valores nos dá la probabilidad de rango: 79.4% aproximadamente.
+ El 20.6% de las personas NO tiene un CI entre 85 y 125.
    + Puede ser mayor o menor pero no está dentro de este rango.

#### Rango de CI o IQ normal o mas frecuente
b. Calcule la probabilidad de que un adulto seleccionado al azar tenga un CI entre 90 y 110 (denominado rango normal).

```python
# b. Probabilidad de que una persona tenga un CI entre 90 y 110, es decir dentro de un rango normal.
min = 90
max = 110

print(f"media: {mean}, desv.std.: {sigm} , sesgo: {skew}, kurtosis: {kurt})")

p1= norm.cdf(max, loc=mean, scale=sigm)
p0= norm.cdf(min, loc=mean, scale=sigm)
p = p1-p0

print(f"Probabilidad de encontrar una persona con un CI entre {min} y {max}: {p}")
distr_prob(norm(loc=mean, scale=sigm), a=min, b=max)

print(f"Probabilidad acumulada {min}: {p0}")
distr_prob(norm(loc=mean, scale=sigm), b=min)

print(f"Probabilidad acumulada inversa {max}: {norm.sf(max, loc=mean, scale=sigm)}")
distr_prob(norm(loc=mean, scale=sigm), a=max)
```

**Interpretación**
+ Apenas un poco menos la mitad de la población (49.5%) tiene un coeficiente intelectual entre 90 y 100.
+ Un poco más de la mitad de la población (50.5%) tiene un CI fuera del rango entre 90 y 110.
    + El 25.25% de la población tiene un CI o IQ menor a 90.
    + El 25.25% de la población tiene un CI O IQ mayor que 110.

#### Rango de CI o IQ normal brillante.
c. Calcule la probabilidad de que un adulto seleccionado al azar tenga un CI entre 110 y 120 (denominado rango de mente brillante).

```python
# b. Probabilidad de que una persona tenga un CI entre 110 y 120, es decir dentro de un rango normal.
min = 110
max = 120

print(f"media: {mean}, desv.std.: {sigm} , sesgo: {skew}, kurtosis: {kurt})")

p1= norm.cdf(max, loc=mean, scale=sigm)
p0= norm.cdf(min, loc=mean, scale=sigm)
p = p1-p0

print(f"Probabilidad de encontrar una persona con un CI entre {min} y {max}: {p}")
distr_prob(norm(loc=mean, scale=sigm), a=min, b=max)

print(f"Probabilidad de encontrar una persona con un CI fuera de {min} y {max}: {1-p}")
distr_prob(norm(loc=mean, scale=sigm), a=min, b=max, between=False)

print(f"Probabilidad acumulada {min}: {p0}")
distr_prob(norm(loc=mean, scale=sigm), b=min)

print(f"Probabilidad acumulada inversa {max}: {norm.sf(max, loc=mean, scale=sigm)}")
distr_prob(norm(loc=mean, scale=sigm), a=max)
```

**Interpretación**

#### Intervalos de confianza
+ Comprobemos los porcentajes de probabilidad de de los intervalos de confianza.
+ 1 La regla del intervalo ofrece intervalos de confianza aproximados:
    + 1. IC 68%. Está delimitado el 68% probabilidad entre -1 desvío y +1 desvío estandard.
    + 2. IC 95%. Está delimitado por el 95% de probabilidad entre -2 desvíos y + dos desvíos estandard.
    + 3. IC 99%. Está delimitado por el 99% de probabilidad entre +-2.575 desvíos estandard.
    + 4. IC 99.7%. Está delimitado por el 99.7% de probabilidad entre +-3 desvíos estandard.
+ 2 La regla del intervalo ofrece tres intervalos más precisos:
    + 1. IC 68.3% para 𝞂 = +-1
    + 2. IC 95% para 𝞂 = +-1.96
    + 3. IC 99% para 𝞂 = +-2.575
    + 4. IC 99.7% para 𝞂 = +-3

```python
print("Intervalo para un nivel de confianza del 68%")
inter_confi(norm(loc=mean, scale=sigm), nc=0.68)
print("Intervalo para un nivel de confianza del 95%")
inter_confi(norm(loc=mean, scale=sigm), nc=0.95)
print("Intervalo para un nivel de confianza del 99%")
inter_confi(norm(loc=mean, scale=sigm), nc=0.99)
print("Intervalo para un nivel de confianza del 99,7%")
inter_confi(norm(loc=mean, scale=sigm), nc=0.997)
```
