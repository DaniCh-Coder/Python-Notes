## Distribución muestral de medias
Supongamos que tenemos una población con una distribución uniforme y queremos estudiar la distribución muestral de la media muestral.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una población uniforme
np.random.seed(42)
poblacion = np.random.uniform(0, 10, 10000)

# Definir el tamaño de las muestras y el número de muestras
tamaño_muestra = 30
n_muestras = 1000

# Calcular las medias muestrales
medias_muestrales = []
for _ in range(n_muestras):
    muestra = np.random.choice(poblacion, tamaño_muestra)
    medias_muestrales.append(np.mean(muestra))

# Gráfico de la distribución muestral de la media
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(medias_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Función de densidad de probabilidad teórica (normal)
media_poblacional = np.mean(poblacion)
desviacion_estandar_poblacional = np.std(poblacion) / np.sqrt(tamaño_muestra)
x = np.linspace(media_poblacional - 4*desviacion_estandar_poblacional, media_poblacional + 4*desviacion_estandar_poblacional, 100)
pdf = (1 / (desviacion_estandar_poblacional * np.sqrt(2 * np.pi))) * np.exp(- (x - media_poblacional)**2 / (2 * desviacion_estandar_poblacional**2))
ax.plot(x, pdf, 'r-', lw=2, label='PDF Teórica')

# Etiquetas y título
ax.set_xlabel('Media Muestral')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Muestral de la Media Muestral')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Histograma de las Medias Muestrales:

+ El histograma muestra la frecuencia de las medias muestrales calculadas a partir de 1000 muestras de tamaño 30 extraídas de la población uniforme.
+ Podemos observar que la distribución muestral de la media se aproxima a una distribución normal, gracias al teorema del límite central.

Función de Densidad de Probabilidad Teórica:
+ La línea roja representa la PDF teórica de la distribución normal que se espera según el teorema del límite central.
+ La superposición de la PDF teórica sobre el histograma muestra que la distribución muestral de la media sigue aproximadamente una distribución normal.

## Distribución Muestral de Proporciones
Supongamos que estamos interesados en estimar la proporción de personas en una ciudad que prefieren el transporte público sobre el transporte privado. 
+ Sabemos que en la población general, aproximadamente el 60% de las personas prefieren el transporte público. 
+ Queremos estudiar la distribución muestral de la proporción utilizando muestras de tamaño n=100.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la población
poblacion_proporcion = 0.6  # Proporción de la población que prefiere el transporte público
tamaño_muestra = 100  # Tamaño de las muestras
n_muestras = 1000  # Número de muestras

# Generar la población
poblacion = np.random.choice([0, 1], size=10000, p=[1-poblacion_proporcion, poblacion_proporcion])

# Calcular las proporciones muestrales
proporciones_muestrales = []
for _ in range(n_muestras):
    muestra = np.random.choice(poblacion, tamaño_muestra)
    proporciones_muestrales.append(np.mean(muestra))

# Gráfico de la distribución muestral de proporciones
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(proporciones_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Función de densidad de probabilidad teórica (normal)
media_poblacional = poblacion_proporcion
desviacion_estandar_muestral = np.sqrt(poblacion_proporcion * (1 - poblacion_proporcion) / tamaño_muestra)
x = np.linspace(media_poblacional - 4*desviacion_estandar_muestral, media_poblacional + 4*desviacion_estandar_muestral, 100)
pdf = (1 / (desviacion_estandar_muestral * np.sqrt(2 * np.pi))) * np.exp(- (x - media_poblacional)**2 / (2 * desviacion_estandar_muestral**2))
ax.plot(x, pdf, 'r-', lw=2, label='PDF Teórica')

# Etiquetas y título
ax.set_xlabel('Proporción Muestral')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Muestral de Proporciones')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Histograma de las Proporciones Muestrales:
+ El histograma muestra la frecuencia de las proporciones muestrales calculadas a partir de 1000 muestras de tamaño 100 extraídas de la población.
+ Podemos observar que la distribución muestral de la proporción se aproxima a una distribución normal, debido al teorema del límite central.

Función de Densidad de Probabilidad Teórica:
+ La línea roja representa la PDF teórica de la distribución normal que se espera según el teorema del límite central.
+ La superposición de la PDF teórica sobre el histograma muestra que la distribución muestral de la proporción sigue aproximadamente una distribución normal.

## Distribución muestral de varianzas
Supongamos que tenemos una población normalmente distribuida con una media 𝜇 = 50  y una desviación estándar 𝜎 = 10. 
+ Queremos estudiar la distribución muestral de la varianza utilizando muestras de tamaño 𝑛 = 30.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Parámetros de la población
mu = 50  # Media de la población
sigma = 10  # Desviación estándar de la población
tamaño_muestra = 30  # Tamaño de las muestras
n_muestras = 1000  # Número de muestras

# Generar una población normal
np.random.seed(42)
poblacion = np.random.normal(mu, sigma, 10000)

# Calcular las varianzas muestrales
varianzas_muestrales = []
for _ in range(n_muestras):
    muestra = np.random.choice(poblacion, tamaño_muestra)
    varianza_muestral = np.var(muestra, ddof=1)  # ddof=1 para obtener varianza muestral
    varianzas_muestrales.append(varianza_muestral)

# Gráfico de la distribución muestral de varianzas
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(varianzas_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Función de densidad de probabilidad teórica (distribución \(\chi^2\) escalada)
df = tamaño_muestra - 1
x = np.linspace(0, max(varianzas_muestrales), 100)
pdf = chi2.pdf(x * tamaño_muestra / sigma**2, df) * tamaño_muestra / sigma**2
ax.plot(x, pdf, 'r-', lw=2, label='PDF Teórica')

# Etiquetas y título
ax.set_xlabel('Varianza Muestral')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Muestral de la Varianza')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Histograma de las Varianzas Muestrales:
+ El histograma muestra la frecuencia de las varianzas muestrales calculadas a partir de 1000 muestras de tamaño 30 extraídas de la población.
+ Podemos observar la distribución muestral de las varianzas que sigue aproximadamente una distribución χ2 escalada.

Función de Densidad de Probabilidad Teórica (PDF):
+ La línea roja ahora representa correctamente la PDF teórica de la distribución χ2 con grados de libertad igual a n−1 y escalada por σ2.
+ La superposición de la PDF teórica sobre el histograma muestra que la distribución muestral de las varianzas sigue la forma esperada según la distribución χ2.

## Distribución Muestral de Diferencia de Medias Muestrales

Supongamos que queremos comparar las alturas de hombres y mujeres en una población. 
+ Vamos a generar datos de dos poblaciones normalmente distribuidas y estudiar la distribución muestral de la diferencia de medias muestrales.
    + Alturas de Hombres: Media μ1 = 175 cm, Desviación estándar σ1 = 10 cm
    + Alturas de Mujeres: Media μ2 = 165 cm, Desviación estándar σ2 = 8 cm

Queremos estudiar la distribución muestral de la diferencia de medias utilizando muestras de tamaño n=50 para ambos grupos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las poblaciones
mu_hombres = 175  # Media de la altura de hombres
sigma_hombres = 10  # Desviación estándar de la altura de hombres
mu_mujeres = 165  # Media de la altura de mujeres
sigma_mujeres = 8  # Desviación estándar de la altura de mujeres
tamaño_muestra = 50  # Tamaño de las muestras
n_muestras = 1000  # Número de muestras

# Generar poblaciones normales
np.random.seed(42)
poblacion_hombres = np.random.normal(mu_hombres, sigma_hombres, 10000)
poblacion_mujeres = np.random.normal(mu_mujeres, sigma_mujeres, 10000)

# Calcular las diferencias de medias muestrales
diferencias_medias_muestrales = []
for _ in range(n_muestras):
    muestra_hombres = np.random.choice(poblacion_hombres, tamaño_muestra)
    muestra_mujeres = np.random.choice(poblacion_mujeres, tamaño_muestra)
    media_hombres = np.mean(muestra_hombres)
    media_mujeres = np.mean(muestra_mujeres)
    diferencias_medias_muestrales.append(media_hombres - media_mujeres)

# Gráfico de la distribución muestral de la diferencia de medias muestrales
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(diferencias_medias_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Función de densidad de probabilidad teórica (normal)
media_diferencia = mu_hombres - mu_mujeres
desviacion_estandar_diferencia = np.sqrt((sigma_hombres**2 / tamaño_muestra) + (sigma_mujeres**2 / tamaño_muestra))
x = np.linspace(media_diferencia - 4*desviacion_estandar_diferencia, media_diferencia + 4*desviacion_estandar_diferencia, 100)
pdf = (1 / (desviacion_estandar_diferencia * np.sqrt(2 * np.pi))) * np.exp(- (x - media_diferencia)**2 / (2 * desviacion_estandar_diferencia**2))
ax.plot(x, pdf, 'r-', lw=2, label='PDF Teórica')

# Etiquetas y título
ax.set_xlabel('Diferencia de Medias Muestrales')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Muestral de la Diferencia de Medias Muestrales')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Histograma de las Diferencias de Medias Muestrales:

El histograma muestra la frecuencia de las diferencias de medias muestrales calculadas a partir de 1000 muestras de tamaño 50 extraídas de las poblaciones de hombres y mujeres.

Podemos observar que la distribución muestral de la diferencia de medias sigue aproximadamente una distribución normal.

Función de Densidad de Probabilidad Teórica (PDF):

La línea roja representa la PDF teórica de la distribución normal de la diferencia de medias muestrales.

La superposición de la PDF teórica sobre el histograma muestra que la distribución muestral de la diferencia de medias sigue la forma esperada según la distribución normal.

## Distribución Muestral del Coeficiente de Correlación de Pearson
El coeficiente de correlación de Pearson (𝑟) mide la fuerza y dirección de la relación lineal entre dos variables. 
+ Vamos a generar datos para dos variables normalmente distribuidas, calcular el coeficiente de correlación de Pearson para múltiples muestras, 
    + y graficar la distribución muestral de este estadístico.

Supongamos que tenemos dos variables, 𝑋 y 𝑌, con una correlación poblacional 𝜌=0.5.

Generamos datos para 𝑋 y 𝑌 con medias 𝜇𝑋=0  y 𝜇𝑌=0, y desviaciones estándar 𝜎𝑋=1 y 𝜎𝑌=1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de las poblaciones
mu_X = 0  # Media de X
sigma_X = 1  # Desviación estándar de X
mu_Y = 0  # Media de Y
sigma_Y = 1  # Desviación estándar de Y
rho = 0.5  # Correlación poblacional entre X y Y
tamaño_muestra = 30  # Tamaño de las muestras
n_muestras = 1000  # Número de muestras

# Generar datos correlacionados
np.random.seed(42)
cov_matrix = [[sigma_X**2, rho*sigma_X*sigma_Y], [rho*sigma_X*sigma_Y, sigma_Y**2]]
poblacion = np.random.multivariate_normal([mu_X, mu_Y], cov_matrix, 10000)

# Calcular los coeficientes de correlación muestrales
correlaciones_muestrales = []
for _ in range(n_muestras):
    muestra = np.random.choice(np.arange(poblacion.shape[0]), tamaño_muestra, replace=False)
    muestra_X = poblacion[muestra, 0]
    muestra_Y = poblacion[muestra, 1]
    coef_correlacion = np.corrcoef(muestra_X, muestra_Y)[0, 1]
    correlaciones_muestrales.append(coef_correlacion)

# Gráfico de la distribución muestral del coeficiente de correlación de Pearson
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(correlaciones_muestrales, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Etiquetas y título
ax.set_xlabel('Coeficiente de Correlación de Pearson')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución Muestral del Coeficiente de Correlación de Pearson')
ax.legend(['Coeficientes muestrales'])

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**
Histograma de los Coeficientes de Correlación Muestrales:

El histograma muestra la frecuencia de los coeficientes de correlación muestrales calculados a partir de 1000 muestras de tamaño 30 extraídas de las poblaciones de 
𝑋 y 𝑌.

Podemos observar la distribución muestral del coeficiente de correlación, que se centra alrededor de la correlación poblacional 
𝜌=0.5.

# Distribución muestral de t-Student
La distribución t de Student se utiliza comúnmente cuando se comparan medias muestrales y la desviación estándar poblacional es desconocida. 
+ Vamos a aplicar la distribución t de Student para comparar las medias de dos muestras y determinar si existe una diferencia significativa entre ellas.

Ejemplo:
+ Supongamos que estamos investigando si una nueva técnica de estudio mejora los puntajes de los estudiantes en comparación con una técnica tradicional. 
    + Tomamos dos muestras independientes:
+ Grupo A (técnica tradicional): 
    + n1 = 30, 
    + media muestral Xˉ1 = 75, 
    + desviación estándar muestral s1 = 10.
+ Grupo B (nueva técnica): 
    + n2 = 30, 
    + media muestral Xˉ2=80, 
    + desviación estándar muestral s2 = 12

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Datos de las muestras
n1 = 30
mean1 = 75
std1 = 10
n2 = 30
mean2 = 80
std2 = 12

# Calcular el estadístico t
pooled_std = np.sqrt(((n1 - 1) * std1**2 + (n2 - 1) * std2**2) / (n1 + n2 - 2))
t = (mean1 - mean2) / (pooled_std * np.sqrt(1/n1 + 1/n2))
df = n1 + n2 - 2

print(f'Estadístico t: {t:.2f}')
print(f'Grados de libertad: {df}')

# Gráfico de la distribución t de Student
x = np.linspace(-4, 4, 100)
y = stats.t.pdf(x, df)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, 'b-', lw=2, label=f't de Student (df={df})')
ax.axvline(x=t, color='r', linestyle='--', label=f'Estadístico t = {t:.2f}')

# Etiquetas y título
ax.set_xlabel('Valor t')
ax.set_ylabel('Densidad de Probabilidad')
ax.set_title('Distribución t de Student')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Cálculo del Estadístico t:
+ El estadístico t se calcula utilizando la diferencia de las medias muestrales, la desviación estándar combinada (pooled standard deviation) y los tamaños de las muestras.
+ En este ejemplo, el estadístico t es aproximadamente -1.74, lo cual indica cuántas desviaciones estándar combinadas está la diferencia de medias de cero.

Gráfico de la Distribución t de Student:
+ La curva azul representa la distribución t de Student con los grados de libertad calculados.
+ La línea vertical roja muestra la ubicación del estadístico t calculado.
