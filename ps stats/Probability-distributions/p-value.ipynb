```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Datos del ejemplo
df = 14  # Grados de libertad
t_obs = 2.5  # Estadístico observado
alpha = 0.05  # Nivel de significancia

# Valores para la distribución t
x = np.linspace(-4, 4, 500)
y = t.pdf(x, df)

# Crear la figura
plt.figure(figsize=(10, 6))

# Dibujar la curva de distribución t
plt.plot(x, y, label="Distribución t", color='blue')

# Sombrear las áreas críticas (región del valor p)
x_shade_right = np.linspace(t_obs, 4, 500)
x_shade_left = np.linspace(-4, -t_obs, 500)
plt.fill_between(x_shade_right, t.pdf(x_shade_right, df), color='red', alpha=0.5, label='Valor p (cola derecha)')
plt.fill_between(x_shade_left, t.pdf(x_shade_left, df), color='red', alpha=0.5, label='Valor p (cola izquierda)')

# Líneas verticales en t observado
plt.axvline(t_obs, color='red', linestyle='--', label=f"t observado = {t_obs:.2f}")
plt.axvline(-t_obs, color='red', linestyle='--')

# Agregar títulos y etiquetas
plt.title("Representación gráfica del valor p en una prueba t", fontsize=14)
plt.xlabel("Valores t", fontsize=12)
plt.ylabel("Densidad de probabilidad", fontsize=12)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)

# Mostrar la gráfica
plt.show()
```
