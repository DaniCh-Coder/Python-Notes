# Distribución Uniforme
###### https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html#scipy.stats.uniform

```python
import numpy as np
from scipy.stats import uniform
import matplotlib.pyplot as plt
import seaborn.objects as so
import seaborn as sns
```

En la forma estándar, la distribución es uniforme en x=0 e y=1. 
+ Utilizando los parámetros "loc" and "scale y , se obtiene la distribución uniforme en loc y loc+scale.
+ por eso en su forma estandard uniform devuelve los estadisticos de [0,1]

```python
# ver los estadísticos de la distribución uniforme como aparecen en su forma estandard
mean, var, skew, kurt = uniform.stats(moments='mvsk')
mean, var, skew, kurt
```

Es decir que en su forma estandard, la función tiene los valores de:
+ Variable independiente: x variables dentro del rango.
+ Función: f o p siempre vale 1 cualquiera sea el valor de x
+ Función inversa: inv de x es igual a x 



```python
x = [0, 0.25, 0.5, 0.75, 1]
fun = uniform.pdf(x)
acu = uniform.cdf(x)
inv = uniform.sf(x)
print(f"Función (pdf):{fun}.    Acumulada (cdf):{acu}.    Inversa (ppf):{inv}.")
```

+ Utilizando los parámetros "loc" and "scale y , se obtiene la distribución uniforme en loc y loc+scale.

Es sabido que una districución es uniforme dentro de un rango a,b

```python
# Parámetros de la distribución uniforme
a = 0.0  # límite inferior
b = 1  # límite superior
loc = a
scale = b - a

# Generar valores de x
x = np.linspace(a, b, 100)

# Crear la distribución uniforme
dist = uniform(loc, scale)

# Calcular la PDF, CDF u SF
pdf = dist.pdf(x)
acu = dist.cdf(x)
inv = dist.sf(x)
```

#### Gráfico - Distribución de Probabiliad Uniforme 

```python
(
    so.Plot(x,pdf)
    .add(so.Area())
    .label(x="x", y="Prob.PDF")
)
```

#### Gráfico - Distribución de Probabiliad Uniforme Acumulada

```python
(
    so.Plot(x,acu)
    .add(so.Area(color='g'))
    .label(x="x", y="Prob.Acu.F")
)
```

#### Gráfico - Distribución de Probabiliad Uniforme Acumulada Inversa (de derecha a Izquierda)

```python
(
    so.Plot(x,inv)
    .add(so.Area(color='r'))
    .label(x="x", y="Prob.Inv.F")
)
```

#### Dataframe con x, distribución de probabilidad uniforme y acumulada

```python
import pandas as pd
df = pd.DataFrame(
    {
        "x"  : x, 
        "fun": pdf,
        "acu": acu
    }
)
df
```

#### Gráfico - Distribución de Probabiliad Uniforme y Acumulada Superpuestos

```python
# Create a figure and axis with desired size
fig, ax = plt.subplots(figsize=(6, 5))

# Create seaborn line plots
sns.lineplot(x="x", y="fun", data=df, color="blue", label="Uniforme", ax=ax)
sns.lineplot(x="x", y="acu", data=df, color="red", label="Acumulada", ax=ax)
sns.set_style('darkgrid')

# Add filled areas using matplotlib (optional)
ax.fill_between(x=df.x, y1=df.fun, facecolor='grey', alpha=0.5)
ax.fill_between(x=df.x, y1=df.acu, facecolor='red', alpha=0.5)

# Customize the plot (optional)
plt.xlabel("x")
plt.ylabel("Probabilidad")
plt.title("Distribución de probabiliad Uniforme y Acumulada")
plt.grid(True)

# Add legend for both lines and filled areas
plt.legend()

# Show the plot
plt.show()
```
