## Ejemplo: Tiempo de recuperación de enfermedad
Supongamos que estamos modelando el tiempo hasta la recuperación de una enfermedad. 
+ Los pacientes tienen un 20% de probabilidad de recuperarse completamente después de cada tratamiento. 
+ Queremos modelar el número de tratamientos necesarios hasta que un paciente se recupere completamente, y estamos interesados en los primeros 5 pacientes que se recuperan.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parámetros de la distribución binomial negativa
r = 5  # Número de éxitos (recuperaciones)
p = 0.2  # Probabilidad de éxito en cada tratamiento

# Simulación del número de tratamientos hasta que se recuperen 5 pacientes
n_simulaciones = 1000  # Número de simulaciones
simulaciones = nbinom.rvs(r, p, size=n_simulaciones)

# Gráfico de la distribución binomial negativa
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(0, max(simulaciones) + 1)
pmf = nbinom.pmf(x, r, p)

# Histogramas y PMF
ax.hist(simulaciones, bins=x, density=True, alpha=0.6, color='g', edgecolor='black', label='Simulaciones')
ax.plot(x, pmf, 'bo', ms=8, label='PMF Teórica')
ax.vlines(x, 0, pmf, colors='b', lw=2)

# Etiquetas y título
ax.set_xlabel('Número de Tratamientos hasta 5 Recuperaciones')
ax.set_ylabel('Probabilidad')
ax.set_title('Distribución Binomial Negativa: Tiempo hasta la Recuperación de la Enfermedad')
ax.legend()

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

**Interpretación del Resultado**

Simulación de Datos:

Generamos datos simulados del número de tratamientos necesarios hasta que 5 pacientes se recuperen completamente utilizando la distribución binomial negativa.

Realizamos 1000 simulaciones para obtener una muestra representativa.

Gráfico de la Distribución:

Creamos un gráfico que muestra tanto el histograma de las simulaciones como la función de masa de probabilidad (PMF) teórica de la distribución binomial negativa.

El eje 
𝑥
 representa el número de tratamientos necesarios.

El eje 
𝑦
 representa la probabilidad de cada número de tratamientos.

Análisis del Gráfico:

El histograma muestra la frecuencia de las simulaciones, representando cuántos tratamientos se necesitaron en cada caso para que 5 pacientes se recuperen.

Los puntos y las líneas verticales en azul muestran la PMF teórica de la distribución binomial negativa.

Podemos observar que la mayoría de los pacientes se recuperan después de un número relativamente bajo de tratamientos, con la probabilidad disminuyendo a medida que aumenta el número de tratamientos necesarios.
